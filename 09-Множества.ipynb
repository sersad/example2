### –£–¥–∞–ª–∏—Ç—å —ç–ª–µ–º–µ–Ω—Ç –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –º–æ–∂–Ω–æ —Ç—Ä–µ–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:

–º–µ—Ç–æ–¥ `.remove()` ‚Äî —É–¥–∞–ª—è–µ—Ç —ç–ª–µ–º–µ–Ω—Ç –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é; –µ—Å–ª–∏ —Ç–∞–∫–æ–≥–æ –Ω–µ—Ç ‚Äî –æ—à–∏–±–∫–∞ KeyError;

## –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ `.union()`
–ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º, —á—Ç–æ –≤—ã —Å–æ —Å–≤–æ–∏–º –ª—É—á—à–∏–º –¥—Ä—É–≥–æ–º —Ö–æ—Ç–∏—Ç–µ —É—Å—Ç—Ä–æ–∏—Ç—å –≤–µ—á–µ—Ä–∏–Ω–∫—É, –Ω–∞ –∫–æ—Ç–æ—Ä—É—é —Ö–æ—Ç–∏—Ç–µ –ø–æ–∑–≤–∞—Ç—å –≤–∞—à–∏—Ö –¥—Ä—É–∑–µ–π –∏ –¥—Ä—É–∑–µ–π –¥—Ä—É–≥–∞ (–ø—Ä–∏ —ç—Ç–æ–º –≤—ã –º–æ–∂–µ—Ç–µ –¥—Ä—É–∂–∏—Ç—å —Å –æ–¥–Ω–∏–º–∏ –∏ —Ç–µ–º–∏ –∂–µ –ª—é–¥—å–º–∏). –°–∞–º—ã–π –ø—Ä–æ—Å—Ç–æ–π —Å–ø–æ—Å–æ–± –ø–æ–ª—É—á–∏—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –≥–æ—Å—Ç–µ–π ‚Äî –æ–ø–µ—Ä–∞—Ü–∏—è –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –º–Ω–æ–∂–µ—Å—Ç–≤. –î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–¥–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –º–µ—Ç–æ–¥–æ–º `.union()` –∏–ª–∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–º `|`.
<img src="img/set-01.jpg" width=300 height=400 />

```python
my_friends = {"–õ–µ–Ω–∞", "–ú–∞—à–∞", "–û–ª–µ–≥"}
friend_friends = {"–ü–µ—Ç—è", "–ú–∞—à–∞", "–ú–∞—Ä–∏–Ω–∞"}
guests = my_friends.union(friend_friends)  # {"–õ–µ–Ω–∞", "–ú–∞—à–∞", "–û–ª–µ–≥", "–ü–µ—Ç—è", "–ú–∞—Ä–∏–Ω–∞"}
print(guests)
```

```python
guests = my_friends | friend_friends
print(guests)
```

## –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ `.intersection()`
–•–æ—Ç–∏—Ç–µ —É–∑–Ω–∞—Ç—å, —á—Ç–æ –æ–±—â–µ–≥–æ —É –¥–≤—É—Ö –º–Ω–æ–∂–µ—Å—Ç–≤? –¢—É—Ç –Ω–∞ –ø–æ–º–æ—â—å –±–µ–∂–∏—Ç –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ. –î–ª—è –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –º–Ω–æ–∂–µ—Å—Ç–≤ –Ω–∞–¥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ `.intersection()` –∏–ª–∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä `&`.

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
```

## —Å–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å `.symmetric_difference()`
–î–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –µ—Å—Ç—å –≤ –∫–∞–∂–¥–æ–º –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤, –Ω–æ –Ω–µ –≤ –æ–±–æ–∏—Ö —Å—Ä–∞–∑—É, –Ω—É–∂–Ω–∞ —Å–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å ‚Äî –º–µ—Ç–æ–¥ `.symmetric_difference()` –∏–ª–∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä `^`.

```python
dev_habits = {"—Ä–∞–±–æ—Ç–∞—Ç—å –ø–æ –Ω–æ—á–∞–º", "–ø–∏—Ç—å –∫–æ—Ñ–µ", 
              "–ø—Ä–æ–∫—Ä–∞—Å—Ç–∏–Ω–∏—Ä–æ–≤–∞—Ç—å", "–ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞—Ç—å –≤–µ—Ç–∫–∏", "—Å–ª—É—à–∞—Ç—å lo-fi"}
cat_habits = {"—Å–ø–∞—Ç—å –¥–Ω—ë–º", "–ø—Ä–æ–∫—Ä–∞—Å—Ç–∏–Ω–∏—Ä–æ–≤–∞—Ç—å", "–∏–≥—Ä–∞—Ç—å –≤ –∫–æ—Ä–æ–±–∫–µ",
              "—Å–º–æ—Ç—Ä–µ—Ç—å –≤ —Å—Ç–µ–Ω—É"}

unique_habits = dev_habits ^ cat_habits  # –ù–µ –ø–æ–ø–∞–¥—ë—Ç –ø—Ä–æ–∫—Ä–∞—Å—Ç–∏–Ω–∞—Ü–∏—è, –∫–∞–∫ –æ–±—â–∞—è –ø—Ä–∏–≤—ã—á–∫–∞.
print(unique_habits)
```
