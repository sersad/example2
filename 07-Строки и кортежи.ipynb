### –î–ª—è –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è –ø–æ–∑–∏—Ü–∏–∏ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤–Ω—É—Ç—Ä–∏ –∫–æ–ª–ª–µ–∫—Ü–∏–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –ø–æ–Ω—è—Ç–∏–µ –∏–Ω–¥–µ–∫—Å. –ö–∞–∫ –∏ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –≤–æ –≤—Å–µ—Ö —è–∑—ã–∫–∞—Ö –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è, –Ω—É–º–µ—Ä–∞—Ü–∏—è –∏–¥—ë—Ç —Å –Ω—É–ª—è.

### –° –∏–Ω–¥–µ–∫—Å–∞–º–∏ –Ω–∞–¥–æ –±—ã—Ç—å –∞–∫–∫—É—Ä–∞—Ç–Ω—ã–º–∏: –µ—Å–ª–∏ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –ø–æ–ª—É—á–∏—Ç—å –±—É–∫–≤—É –ø–æ –∏–Ω–¥–µ–∫—Å—É, –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–µ—Ç –≤ —Å—Ç—Ä–æ–∫–µ, —Ç–æ –ø–æ–ª—É—á–∏—Ç–µ –æ—à–∏–±–∫—É IndexError.

```python
print(''[100500])  # IndexError ‚Äî –≤ –ø—É—Å—Ç–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–µ—Ç –±—É–∫–≤—ã —Å —Ç–∞–∫–∏–º –∏–Ω–¥–µ–∫—Å–æ–º (–Ω–∏–∫–∞–∫–æ–π –Ω–µ—Ç).
```

## üéØ –ú—ã —Ç–µ–±—è –¥–æ—Å—Ç–∞–Ω–µ–º —Ç–∞–∫ –∏–ª–∏ –∏–Ω–∞—á–µ
–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ, —á—Ç–æ –º—ã –¥–µ–ª–∏–º —Å—Ç—Ä–æ–∫–æ–≤—É—é –ø–∏—Ü—Ü—É –º–µ–∂–¥—É —á–µ—Ç—ã—Ä—å–º—è –¥—Ä—É–∑—å—è–º–∏:

```python
pizza_str = "–ü–∏—Ü—Ü–∞ –ø–µ–ø–ø–µ—Ä–æ–Ω–∏"
idx = 0  # –ü–µ—Ä–≤—ã–π –∫—É—Å–æ—á–µ–∫ —Å –∏–Ω–¥–µ–∫—Å–æ–º 0.
for piece in pizza_str:
    match idx % 4:
        case 0:
            friend = "–õ–µ–æ–Ω–∞—Ä–¥–æ"
        case 1:
            friend = "–†–∞—Ñ–∞—ç–ª—å"
        case 2:
            friend = "–ú–∏–∫–µ–ª–∞–Ω–¥–∂–µ–ª–æ"
        case 3:
            friend = "–î–æ–Ω–∞—Ç–µ–ª–ª–æ"
    print(f"{friend} –ø–æ–ª—É—á–∞–µ—Ç –∫—É—Å–æ—á–µ–∫ {idx + 1} —Å –±—É–∫–≤–æ–π {piece}.")
    idx += 1
```

–ï—Å—Ç—å —Å–ø–æ—Å–æ–± –µ—â—ë –ª—É—á—à–µ. Python –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –∏ —Ö–æ–¥–∏—Ç—å –ø–æ –Ω–æ–º–µ—Ä—É —ç–ª–µ–º–µ–Ω—Ç–∞, –∏ –ø–æ–ª—É—á–∞—Ç—å —Å–∞–º —ç–ª–µ–º–µ–Ω—Ç. –î–ª—è —ç—Ç–æ–≥–æ –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π `enumerate()`. –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∫–æ–ª–ª–µ–∫—Ü–∏—é –∏ –Ω–æ–º–µ—Ä –ø–µ—Ä–≤–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞. –ï—Å–ª–∏ –Ω–æ–º–µ—Ä –Ω–µ –ø–µ—Ä–µ–¥–∞—Ç—å, —Ç–æ –Ω—É–º–µ—Ä–∞—Ü–∏—è –±—É–¥–µ—Ç –∏–¥—Ç–∏ —Å –Ω—É–ª—è.

```python
pizza_str = "–ü–æ—Å–ª–µ–¥–Ω—è—è –ø–∏—Ü—Ü–∞ ‚Äî —Å –∞–Ω—á–æ—É—Å–∞–º–∏"
for idx, piece in enumerate(pizza_str):  # –¢—É—Ç –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –º–∞–≥–∏—è —Ä–∞—Å–ø–∞–∫–æ–≤–∫–∏ –ø—Ä—è–º–æ –≤ –∑–∞–≥–æ–ª–æ–≤–∫–µ —Ü–∏–∫–ª–∞.
    match idx % 4:
        case 0:
            friend = "–õ–µ–æ–Ω–∞—Ä–¥–æ"
        case 1:
            friend = "–†–∞—Ñ–∞—ç–ª—å"
        case 2:
            friend = "–ú–∏–∫–µ–ª–∞–Ω–¥–∂–µ–ª–æ"
        case 3:
            friend = "–î–æ–Ω–∞—Ç–µ–ª–ª–æ"
    print(f"{friend} –ø–æ–ª—É—á–∞–µ—Ç –∫—É—Å–æ—á–µ–∫ {idx + 1} —Å –±—É–∫–≤–æ–π {piece}.")
```

–í —Ü–µ–ª–æ–º –≤—Å–µ —Ç—Ä–∏ –≤–∞—Ä–∏–∞–Ω—Ç–∞ –≤—ã—à–µ –∏–º–µ—é—Ç –ø—Ä–∞–≤–æ –Ω–∞ –∂–∏–∑–Ω—å, –∏ –Ω–∏–∫—Ç–æ –Ω–∞ –≤–∞—Å –∫–æ—Å–æ –Ω–µ –ø–æ—Å–º–æ—Ç—Ä–∏—Ç, –µ—Å–ª–∏ –≤—ã –±—É–¥–µ—Ç–µ –ø—Ä–∏–¥–µ—Ä–∂–∏–≤–∞—Ç—å—Å—è –≤—Å–µ–≥–¥–∞ –∫–∞–∫–æ–≥–æ-—Ç–æ –æ–¥–Ω–æ–≥–æ —Å—Ç–∏–ª—è –Ω–∞–ø–∏—Å–∞–Ω–∏—è –ø–æ–¥–æ–±–Ω—ã—Ö –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤. –ù–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –Ω–µ–ø—Ä–æ—à–µ–Ω–æ–≥–æ —Å–æ–≤–µ—Ç–∞ –Ω–∞–ø–æ–º–Ω–∏–º:

- –≤–∞–∂–Ω—ã —Ç–æ–ª—å–∫–æ —Å–∞–º–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –∫–æ–ª–ª–µ–∫—Ü–∏–∏ ‚Äî –ø—Ä–æ—Ö–æ–¥–∏–º—Å—è –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º;
- –≤–∞–∂–Ω–∞ —Å–∫–æ—Ä–µ–µ –ø–æ–∑–∏—Ü–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ ‚Äî `range()`;
- –Ω—É–∂–Ω–æ –∏ —Ç–æ –∏ —Ç–æ ‚Äî `enumerate()`.

### –ó–∞–¥–∞—á–∞ ¬´–ö—Ç–æ –Ω–∞ —Å–≤–µ—Ç–µ‚Ä¶¬ª

### –ó–∞–¥–∞—á–∞ ¬´–ù–∞—Ä—É—à–∏—Ç–µ–ª—å–Ω–∏—Ü–∞¬ª

## üß© –ù–∞ –º–µ–ª–∫–∏–µ –∫—É—Å–æ—á–∫–∏

`line[idx]` ‚Äî –ø–æ–ª—É—á–∏—Ç—å –æ–¥–∏–Ω —ç–ª–µ–º–µ–Ω—Ç (–±—É–∫–≤—É) —Å –∏–Ω–¥–µ–∫—Å–æ–º idx (—ç—Ç–æ –º—ã —É–∂–µ –∑–Ω–∞–µ–º).

`line[start:stop]` ‚Äî –ø–æ–ª—É—á–∏—Ç—å –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å –ø–æ–∑–∏—Ü–∏–∏ start –¥–æ –ø–æ–∑–∏—Ü–∏–∏ stop –Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ, —Ç. –µ. —Å—Ç—Ä–æ–∫—É, —Å–æ–¥–µ—Ä–∂–∞—â—É—é –≤—Å–µ –±—É–∫–≤—ã –∏—Å—Ö–æ–¥–Ω–æ–π, —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω—ã–µ –ø–æ —ç—Ç–∏–º –∏–Ω–¥–µ–∫—Å–∞–º.

`line[start:stop:step]` ‚Äî –ø–æ–ª—É—á–∏—Ç—å –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å –ø–æ–∑–∏—Ü–∏–∏ start –¥–æ –ø–æ–∑–∏—Ü–∏–∏ stop –Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ —Å —à–∞–≥–æ–º step. –®–∞–≥, –∫–∞–∫ –∏ –≤ range(), –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º, —Ç–æ–≥–¥–∞ –∏–Ω–¥–µ–∫—Å –Ω–∞—á–∞–ª–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –±–æ–ª—å—à–µ –∏–Ω–¥–µ–∫—Å–∞ –∫–æ–Ω—Ü–∞.


–ö—Ä–æ–º–µ —Ç–æ–≥–æ, —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∏ —Å–æ–∫—Ä–∞—â–µ–Ω–Ω—ã–µ –∑–∞–ø–∏—Å–∏:

`line[:stop]` ‚Äî –µ—Å–ª–∏ –Ω–µ —É–∫–∞–∑–∞–Ω–æ –Ω–∞—á–∞–ª–æ, —Ç–æ –æ—Ç –Ω–∞—á–∞–ª–∞ –¥–æ –Ω—É–∂–Ω–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞ –Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

`line[start:]` ‚Äî –Ω–µ —É–∫–∞–∑–∞–Ω –∫–æ–Ω–µ—Ü ‚Äî –¥–æ –∫–æ–Ω—Ü–∞ —Å—Ç—Ä–æ–∫–∏/–∫–æ–ª–ª–µ–∫—Ü–∏–∏.

`line[:]` ‚Äî —Å—Ç—Ä–æ–∫–∞ –æ—Ç –Ω–∞—á–∞–ª–∞ –¥–æ –∫–æ–Ω—Ü–∞ (–ø–æ–ª–Ω–∞—è –∫–æ–ø–∏—è ‚Äî –¥–ª—è —Å—Ç—Ä–æ–∫ –Ω–µ –æ—Å–æ–±–æ –ø–æ–ª–µ–∑–Ω–æ, –Ω–æ –ø—Ä–∏–≥–æ–¥–∏—Ç—Å—è –¥–ª—è –¥—Ä—É–≥–æ–π –∫–æ–ª–ª–µ–∫—Ü–∏–∏).

`line[::step]` ‚Äî –æ—Ç –Ω–∞—á–∞–ª–∞ –¥–æ –∫–æ–Ω—Ü–∞ —Å —à–∞–≥–æ–º.

`line[::-1]` ‚Äî –±—É–∫–≤—ã —Å—Ç—Ä–æ–∫–∏ / —ç–ª–µ–º–µ–Ω—Ç—ã –∫–æ–ª–ª–µ–∫—Ü–∏–∏ –≤ –æ–±—Ä–∞—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.

```python
text = "—É–∂–∞—Å–æ–ø–ª–∏–≤–æ–≥–æ–≤–∏–ª–ø–æ—Å–∞–∂—É"  # –•–æ—Ä–æ—à–æ, —á—Ç–æ –Ω–µ –ø–∏—Ç–æ–Ω–∞.
```

```python
print(text[2:7])    # –∞—Å–æ–ø–ª (–∏–Ω–¥–µ–∫—Å—ã 2, 3, 4, 5, 6).
```

```python
print(text[::2])    # —É–∞–æ–ª–≤–≥–≤–ª–æ–∞—É (–∫–∞–∂–¥—ã–π –≤—Ç–æ—Ä–æ–π —Å–∏–º–≤–æ–ª).
```

```python
print(text[::-1])   # —É–∂–∞—Å–æ–ø–ª–∏–≤–æ–≥–æ–≤–∏–ª–ø–æ—Å–∞–∂—É ‚Äî –ø–∞–ª–∏–Ω–¥—Ä–æ–º!
```

### –ü—Ä–µ–ª–µ—Å—Ç—å —Å—Ä–µ–∑–æ–≤ –≤ —Ç–æ–º, —á—Ç–æ –æ–Ω–∏ –Ω–µ –≥–µ–Ω–µ—Ä–∏—Ä—É—é—Ç –æ—à–∏–±–∫—É `IndexError`. –í —Å—Ä–µ–∑ –ø–æ–ø–∞–¥–∞–µ—Ç —Ç–æ, —á—Ç–æ –º–æ–∂–Ω–æ –∑–∞–±—Ä–∞—Ç—å –∏–∑ —Å—Ç—Ä–æ–∫–∏, –¥–∞–∂–µ –µ—Å–ª–∏ –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –±—É–¥–µ—Ç –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞.

```python
print("–°—Ä–µ–∑ –¥–æ –º–∏–ª–ª–∏–æ–Ω–∞"[:1_000_000])  # -> –°—Ä–µ–∑ –¥–æ –º–∏–ª–ª–∏–æ–Ω–∞.
```

```python
print(""[20:30:1_000])  # -> –ü—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞, –∏ –Ω–∏–∫–∞–∫–æ–π –æ—à–∏–±–∫–∏.
```

–î–æ–ø—É—Å—Ç–∏–º, –º—ã —Ö–æ—Ç–∏–º –ø–æ–ª—É—á–∏—Ç—å –ª—É—á—à–µ–≥–æ —É—á–µ–Ω–∏–∫–∞ –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –∫—Ä–∏—Ç–µ—Ä–∏—è–º:

- –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª,
- –±–∞–ª–ª –ø–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–µ,
- –±–∞–ª–ª –ø–æ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ,
- –ø—Ä–∏ –ø—Ä–æ—á–∏—Ö —Ä–∞–≤–Ω—ã—Ö ‚Äî —Ç–æ–≥–æ, —á—å—è —Ñ–∞–º–∏–ª–∏—è —Å—Ç–æ–∏—Ç —Ä–∞–Ω—å—à–µ –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É.

–û–±—ä–µ–¥–∏–Ω–∏–º —ç—Ç—É –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –≤ –∫–æ—Ä—Ç–µ–∂–∏ –≤ —Ç–æ–º –ø–æ—Ä—è–¥–∫–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞–º –≤–∞–∂–Ω—ã –∫—Ä–∏—Ç–µ—Ä–∏–∏.

```python
lastname1 = "–•–∞–π—Ä—É–ª–ª–∏–Ω"
lastname2 = "–¶–∏—Ü—é—Ä–∞"
lastname3 = "–ü–æ–ª—Ç–∞–ø–æ–≤"
# –ï—Å–ª–∏ –ø—Ä–æ—Å—Ç–æ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ —Ñ–∞–º–∏–ª–∏–∏, —Ç–æ –ø–æ–±–µ–¥–∏—Ç –ø—Ä–∏ –ø—Ä–æ—á–∏—Ö —Ä–∞–≤–Ω—ã—Ö —Ç–æ—Ç,
# —á—å—è —Ñ–∞–º–∏–ª–∏—è –ø–æ–∑–∂–µ –∞–ª—Ñ–∞–≤–∏—Ç—É, –ø–æ—ç—Ç–æ–º—É –±–µ—Ä—ë–º –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ–¥–∞ –ø–µ—Ä–≤–æ–π –±—É–∫–≤—ã.
student1 = (4.9, 4.8, 5.0, -ord(lastname1[0]), lastname1)
student2 = (4.9, 4.7, 5.0, -ord(lastname2[0]), lastname2)
student3 = (4.9, 4.8, 5.0, -ord(lastname3[0]), lastname3)
print(max(student1, student2, student3)[-1])
```

## üõ∏ –ê –≤—ã –∏–∑ –∫–∞–∫–æ–π —Å–∏—Å—Ç–µ–º—ã?

**–î–≤–æ–∏—á–Ω–∞—è** ‚Äî –¥–ª—è –Ω–∞—Å—Ç–æ—è—â–∏—Ö –≥–∏–∫–æ–≤ –∏ –º–∏–∫—Ä–æ–∫–æ–Ω—Ç—Ä–æ–ª–ª–µ—Ä–æ–≤ (–∫–∞–∫ –º—ã –≥–æ–≤–æ—Ä–∏–ª–∏ –Ω–∞ –ø–µ—Ä–≤–æ–º —É—Ä–æ–∫–µ, —ç—Ç–æ —Ä–æ–¥–Ω–æ–π —è–∑—ã–∫ –ø—Ä–æ—Ü–µ—Å—Å–æ—Ä–æ–≤). –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è bin().

**–í–æ—Å—å–º–µ—Ä–∏—á–Ω–∞—è** ‚Äî –µ—Å–ª–∏ –≤—ã –∑–∞–±–ª—É–¥–∏–ª–∏—Å—å –≤ —Å—Ç–∞—Ä—ã—Ö UNIX-–º–∞–Ω—É–∞–ª–∞—Ö (oct()).

**–î–µ—Å—è—Ç–∏—á–Ω–∞—è** ‚Äî –¥–ª—è –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö –ª—é–¥–µ–π. –¢–∞–∫ —á–∏—Å–ª–∞ –≤ Python –≤—ã–≥–ª—è–¥—è—Ç –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é.

**–®–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–∞—è** ‚Äî –∫–æ–≥–¥–∞ —Ö–æ—á–µ—à—å –≤—ã–≥–ª—è–¥–µ—Ç—å —É–º–Ω—ã–º –∏ —Ä–∞–±–æ—Ç–∞–µ—à—å —Å —Ü–≤–µ—Ç–∞–º–∏ –∏–ª–∏ –∞–¥—Ä–µ—Å–∞–º–∏ –≤ –ø–∞–º—è—Ç–∏ (hex()).

```python
number = 42  

# –î–≤–æ–∏—á–Ω–∞—è —Å–∏—Å—Ç–µ–º–∞ (—Ç–æ–ª—å–∫–æ 0 –∏ 1).  
print(bin(number))  # 0b101010  
# 0b –≤ –Ω–∞—á–∞–ª–µ ‚Äî —ç—Ç–æ –∫–∞–∫ –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: "–û—Å—Ç–æ—Ä–æ–∂–Ω–æ, –∑–¥–µ—Å—å —Ç–æ–ª—å–∫–æ –Ω—É–ª–∏ –∏ –µ–¥–∏–Ω–∏—Ü—ã!"

# –í–æ—Å—å–º–µ—Ä–∏—á–Ω–∞—è (—Ü–∏—Ñ—Ä—ã 0-7).  
print(oct(number))  # 0o52  

# –®–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–∞—è (0-9 –∏ A-F).  
print(hex(number))  # 0x2a  
```

```python
a1 = int('0b1010', 2)  # 10
a2 = int('0o12', 8)    # 10
a3 = int('0xff', 16)   # 255
```

```python
print(a1)
print(a2)
print(a3)
```
