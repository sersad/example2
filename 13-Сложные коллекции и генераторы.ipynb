```python
monday_plans = [
    "–∫–æ—Ñ–µ",
    "–Ω–µ –ø–∞–Ω–∏–∫–æ–≤–∞—Ç—å",
    "–ö–æ—Ñ–µ",
    "—Å–¥–µ–ª–∞—Ç—å –≤–∏–¥, —á—Ç–æ –≤—Å—ë –ø–æ–¥ –∫–æ–Ω—Ç—Ä–æ–ª–µ–º",
    "–ö–û–§–ï"
]
unique_words = {w.lower() for w in monday_plans}
unique_words
```

–¢–∞–∫–∂–µ –º–æ–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å —É—Å–ª–æ–≤–∏—è –ø–æ –≤–∫—É—Å—É. –ù–∞–ø—Ä–∏–º–µ—Ä, —Ö–æ—Ç–∏–º –ø–æ–ª—É—á–∏—Ç—å —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤ –≤–µ—Ä—Ö–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ –±–µ–∑ –ª–∏—à–Ω–∏—Ö –ø—Ä–æ–±–µ–ª–æ–≤, –µ—Å–ª–∏ –¥–ª–∏–Ω–∞ –∏—Å—Ö–æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ —á—ë—Ç–Ω–∞—è: 

```python
brain_mode = [
    "–æ—à–∏–±–∫–∞",
    "   –æ—à–∏–±–∫–∞",
    " —ç–π—Ñ–æ—Ä–∏—è  ",
    "–æ—à–∏–±–∫–∞   ",
    "–ø–µ—Ä–µ–∑–∞–≥—Ä—É–∑–∫–∞   "
]
cleaned = {el.strip().upper() for el in brain_mode if len(el) % 2 == 0}
cleaned
```

## ü¶• –õ–µ–Ω–∏–≤—ã–π –Ω–µ –∑–Ω–∞—á–∏—Ç –±–µ—Å–ø–æ–ª–µ–∑–Ω—ã–π
–¢–∞–∫, –µ—Å–ª–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ —Å–∫–æ–±–∫–∏ –≤ —Å–ø–∏—Å–æ—á–Ω–æ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–∏ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —Ñ–∏–≥—É—Ä–Ω—ã–µ, —Ç–æ –ø–æ–ª—É—á–∏–º –º–Ω–æ–∂–µ—Å—Ç–≤–∞. –ó–Ω–∞—á–∏—Ç –ª–∏ —ç—Ç–æ, —á—Ç–æ, –ø–æ–º–µ–Ω—è–≤ —Å–∫–æ–±–∫–∏ –Ω–∞ –∫—Ä—É–≥–ª—ã–µ, –ø–æ–ª—É—á–∏–º –∫–æ—Ä—Ç–µ–∂? –ù–µ—Ç, –Ω–µ —Ç–∞–∫. –ö–æ—Ä—Ç–µ–∂–∏ ‚Äî –Ω–µ–∏–∑–º–µ–Ω—è–µ–º—ã–µ –∫–æ–ª–ª–µ–∫—Ü–∏–∏, –ø–æ—ç—Ç–æ–º—É –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ–º —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏—Ö —Å–æ–∑–¥–∞—Ç—å –Ω–µ –ø–æ–ª—É—á–∏—Ç—Å—è. –ù–æ, –Ω–∞–ø–∏—Å–∞–≤ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –≤ –∫—Ä—É–≥–ª—ã—Ö —Å–∫–æ–±–∫–∞—Ö, –º—ã —Ç–æ–∂–µ –ø–æ–ª—É—á–∏–º —á—Ç–æ-—Ç–æ –ø–æ–ª–µ–∑–Ω–æ–µ ‚Äî –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –∏, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –Ω–∞ –≤—ã—Ö–æ–¥–µ.

```python
gen = (x * 2 for x in range(1000000))
print(gen)
```

–î–æ–ø—É—Å—Ç–∏–º, –Ω–∞–º –Ω—É–∂–Ω–æ –ø–µ—Ä–µ–±—Ä–∞—Ç—å –º–∏–ª–ª–∏–æ–Ω —á–∏—Å–µ–ª. –•—Ä–∞–Ω–∏—Ç—å –∏—Ö –≤ —Å–ø–∏—Å–∫–µ ‚Äî –º–æ–∂–Ω–æ, –Ω–æ –µ—Å—Ç –ø–∞–º—è—Ç—å. –ê –µ—Å–ª–∏ –º—ã –ø—Ä–æ—Å—Ç–æ —Ö–æ—Ç–∏–º –ø—Ä–æ–π—Ç–∏—Å—å –ø–æ –Ω–∏–º ‚Äî –∑–∞—á–µ–º —Ö—Ä–∞–Ω–∏—Ç—å?

–ì–µ–Ω–µ—Ä–∞—Ç–æ—Ä –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –≤–µ—Å—å —Å—Ä–∞–∑—É –∏ –æ—Ç–¥–∞—ë—Ç —Å–ª–µ–¥—É—é—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ç–æ–ª—å–∫–æ –≤ —Ç–æ–º —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ –µ–≥–æ –ø–æ–ø—Ä–æ—Å—è—Ç. –ó–∞ —ç—Ç–æ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–µ –Ω–∞–∑—ã–≤–∞—é—Ç –ª–µ–Ω–∏–≤—ã–º–∏.

```python
# –°–ø–∏—Å–æ–∫ (–∂–∞–¥–Ω—ã–π).
numbers_list = [x ** 2 for x in range(10000000)]  # –ó–∞–Ω–∏–º–∞–µ—Ç 100 –ú–ë –ø–∞–º—è—Ç–∏.
```

```python
# –ì–µ–Ω–µ—Ä–∞—Ç–æ—Ä (–ª–µ–Ω–∏–≤—ã–π).
numbers_gen = (x ** 2 for x in range(10000000000000000000000000000000))  # –ó–∞–Ω–∏–º–∞–µ—Ç 200 –±–∞–π—Ç.
for n in numbers_gen:
    if n > 100:
        break
    print(n)
```

–î–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å–ª–µ–¥—É—é—â–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ –º–æ–∂–Ω–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –Ω–µ —Ç–æ–ª—å–∫–æ —Ü–∏–∫–ª–æ–º `for`, –Ω–æ –∏ —Ñ—É–Ω–∫—Ü–∏–µ–π `next()`, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–¥–Ω–æ —Å–ª–µ–¥—É—é—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ (–∏–ª–∏, –µ—Å–ª–∏ –≥–æ–≤–æ—Ä–∏—Ç—å –ø—Ä–∞–≤–∏–ª—å–Ω–µ–µ –∏ –±–æ–ª–µ–µ –æ–±—â–æ, –∏–∑ –∏—Ç–µ—Ä–∞—Ç–æ—Ä–∞).

```python
numbers_gen = (x ** 2 for x in range(1, 5))
while a := next(numbers_gen, None):  # None –≤–µ—Ä–Ω—ë—Ç—Å—è, –∫–æ–≥–¥–∞ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–µ –∑–∞–∫–æ–Ω—á–∞—Ç—Å—è.
    print(a)
#     if a > 1000000:
#         break
print(list(numbers_gen))
```

## –î–µ–∫–∞—Ä—Ç–æ–≤–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ

–§—É–Ω–∫—Ü–∏—è product ‚Äî –¥–µ–∫–∞—Ä—Ç–æ–≤–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ. –¢–æ –µ—Å—Ç—å –∫–æ–º–±–∏–Ω–∞—Ü–∏—è –≤—Å–µ–≥–æ —Å–æ –≤—Å–µ–º.

```python
from itertools import product

colors = ['–∫—Ä–∞—Å–Ω—ã–π', '—Å–∏–Ω–∏–π']
sizes = ['S', 'M', 'L']

for combo in product(colors, sizes):
    print(combo)
```

–§—É–Ω–∫—Ü–∏—è –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∏ –±–æ–ª—å—à–µ –¥–≤—É—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:

```python
from itertools import product

for combo in product(range(24), range(60), range(60)):
    print(':'.join(str(x) for x in combo))
```

–ï—Å—Ç—å –Ω–µ–æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä repeat, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª—è–µ—Ç –¥—É–±–ª–∏—Ä–æ–≤–∞—Ç—å –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–µ –∏—Ç–µ—Ä–∏—Ä—É–µ–º—ã–µ –æ–±—ä–µ–∫—Ç—ã –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑. –†–∞–±–æ—Ç–∞–µ—Ç –∫–∞–∫ —Å –æ–¥–Ω–∏–º, —Ç–∞–∫ –∏ —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –∏—Ç–µ—Ä–∞—Ç–æ—Ä–∞–º–∏.

```python
from itertools import product

bits = ['0', '1']
for combo in product(bits, repeat=8):  # (8 –±–∏—Ç == 1 –±–∞–π—Ç)
    print(combo)

start_phrase = ["—Ä–∞–¥—ã –≤–∏–¥–µ—Ç—å –≤–∞—Å", "—Ä–∞—Å—Å–∫–∞–∂–∏—Ç–µ –æ —Å–µ–±–µ", "–Ω–∞–º –≤—Å—ë –ø–æ–Ω—Ä–∞–≤–∏–ª–æ—Å—å"]
end_phrase = ["–Ω–æ —É –Ω–∞—Å —É–∂–µ –µ—Å—Ç—å –∫–∞–Ω–¥–∏–¥–∞—Ç", "–º—ã –≤–∞–º –ø–µ—Ä–µ–∑–≤–æ–Ω–∏–º", "—É–¥–∞—á–∏"]

for phrase in product(start_phrase, end_phrase, repeat=2):
    print("HR –≥–æ–≤–æ—Ä–∏—Ç:", ', '.join(phrase))
```

–§—É–Ω–∫—Ü–∏—è combinations –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∑–∞–¥–∞–Ω–Ω–æ–π –¥–ª–∏–Ω—ã –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π, combinations_with_replacement ‚Äî —Å –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏—è–º–∏.

–ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏

–í—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∫–æ–ª–ª–µ–∫—Ü–∏–∏ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ permutation(). –ú–æ–∂–Ω–æ —É–∫–∞–∑–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–µ; –µ—Å–ª–∏ –Ω–∏—á–µ–≥–æ –Ω–µ —É–∫–∞–∑—ã–≤–∞—Ç—å, —Ç–æ –±—É–¥—É—Ç –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏, —Ä–∞–≤–Ω—ã–µ –¥–ª–∏–Ω–µ –∫–æ–ª–ª–µ–∫—Ü–∏–∏.

```python
from itertools import permutations

for perm in permutations("–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞", 3):
    print(perm)
```

–¶–µ–ø–æ—á–∫–∞ –∏—Ç–µ—Ä–∞—Ç–æ—Ä–æ–≤

–° –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ chain() –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Ü–∞—Ä—å-–∏—Ç–µ—Ä–∞—Ç–æ—Ä –∏–∑ –∫–æ–ª–ª–µ–∫—Ü–∏–∏ –∏—Ç–µ—Ä–∞—Ç–æ—Ä–æ–≤ –ø–æ–º–µ–Ω—å—à–µ.

```python
from itertools import chain

a = [1, 2]
b = [3, 4]
c = [5, 6]

for x in chain(a, b, c, range(7, 11), "hello"):
    print(x)
```

–ë–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª

–§—É–Ω–∫—Ü–∏—è cycle() –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–æ–∑–¥–∞—Ç—å –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª –∏–∑ –ª—é–±–æ–≥–æ –∏—Ç–µ—Ä–∏—Ä—É–µ–º–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞. –ö–∞–∫ —Ç–æ–ª—å–∫–æ –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –∑–∞–∫–∞–Ω—á–∏–≤–∞—é—Ç—Å—è, –º—ã –ø–æ–π–¥—ë–º —Å –Ω–∞—á–∞–ª–∞.

```python
from itertools import cycle

tasks_workflow = [
    "–ø–ª–∞–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ",
    "—Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∞",
    "—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ",
    "–≤–Ω–µ–¥—Ä–µ–Ω–∏–µ",
    "—Ä–µ—Ç—Ä–æ—Å–ø–µ–∫—Ç–∏–≤–∞",
    "–∫–æ—Ñ–µ"
]

lifecycle = cycle(tasks_workflow)
action_points = 100
while action_points:
    action = next(lifecycle)
    if action == "–∫–æ—Ñ–µ":
        print("–í—Ä–µ–º—è –¥–ª—è –∫–æ—Ñ–µ!")
    else:
        print(f"–í—ã–ø–æ–ª–Ω—è–µ–º –∑–∞–¥–∞—á—É: {action}")
    action_points -= 1
```

–°—Ä–µ–∑ –¥–ª—è –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã—Ö –∏—Ç–µ—Ä–∞—Ç–æ—Ä–æ–≤

–ò–Ω–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –∫–∞–∫–∏–µ-—Ç–æ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –∏—Ç–µ—Ä–∏—Ä—É–µ–º–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞. –ù–µ –≤—Å–µ–≥–¥–∞ —ç—Ç–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å –ø–æ–º–æ—â—å—é –ø—Ä–∏–≤—ã—á–Ω–æ–≥–æ –Ω–∞–º —Å—Ä–µ–∑–∞, —Ç–∞–∫ –∫–∞–∫ –∏–Ω–æ–≥–¥–∞ —ç—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ñ—É–Ω–∫—Ü–∏–∏ chain() –∏–ª–∏ –≤–æ–æ–±—â–µ –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π –∏—Ç–µ—Ä–∞—Ç–æ—Ä ‚Äî —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ñ—É–Ω–∫—Ü–∏–∏ cycle(). –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –Ω–∞ –ø–æ–º–æ—â—å –ø—Ä–∏–¥—ë—Ç islice().

```python
from itertools import islice, cycle

for task in islice(cycle(['–ø–æ–µ—Å—Ç—å', '–ø–æ—Å–ø–∞—Ç—å', '–ø–æ–¥—Ä–∞—Ç—å –¥–∏–≤–∞–Ω']), 10, 15):  # –ß—Ç–æ —Ç–∞–º —É –∫–æ—Ç–∏–∫–∞ –≤ —Å–ø–∏—Å–∫–µ –¥–µ–ª –Ω–∞ –ø—É–Ω–∫—Ç–∞—Ö 10‚Äî14?
    print(task)
```

## –°—á—ë—Ç—á–∏–∫

–ü–æ–º–Ω–∏—Ç–µ, —Å–ª–æ–≤–∞—Ä–∏ —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è –ø–æ–¥—Å—á—ë—Ç–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –∫–∞–∫–æ–π-–Ω–∏–±—É–¥—å –¥—Ä—É–≥–æ–π –∫–æ–ª–ª–µ–∫—Ü–∏–∏. –ù–æ –¥–ª—è —ç—Ç–æ–≥–æ –µ—Å—Ç—å –∏ –±–æ–ª–µ–µ –ø–æ–¥—Ö–æ–¥—è—â–∏–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç ‚Äî Counter.

```python
from collections import Counter

orders = ['–ø–∏—Ü—Ü–∞', '–±—É—Ä–≥–µ—Ä', '–ø–∏—Ü—Ü–∞', '—Å—É—à–∏', '–ø–∏—Ü—Ü–∞', '–±—É—Ä–≥–µ—Ä']
counter = Counter(orders)
print(counter)
```

–¢–æ –µ—Å—Ç—å Counter —Ä–µ—à–∞–µ—Ç –≤–ø–æ–ª–Ω–µ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—É—é –∑–∞–¥–∞—á—É, –∫–æ—Ç–æ—Ä—É—é –º–æ–∂–Ω–æ —Ä–µ—à–∏—Ç—å –∏ —Å–ª–æ–≤–∞—Ä—ë–º, –ø—Ä–æ—Å—Ç–æ –Ω–∞–¥–æ –±—É–¥–µ—Ç –Ω–∞–ø–∏—Å–∞—Ç—å –µ—â—ë –Ω–µ–º–Ω–æ–≥–æ –∫–æ–¥–∞. –ò–∑ –∫–ª–∞—Å—Å–Ω–æ–≥–æ: –µ—Å—Ç—å –º–µ—Ç–æ–¥ .most_common(n), –∫–æ—Ç–æ—Ä—ã–π –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç n —Å–∞–º—ã—Ö —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏—Ö—Å—è —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

```python
from collections import Counter

cat_votes = ["–ë–∞—Ä—Å–∏–∫", "–ú—É—Ä–∑–∏–∫", "–ë–∞—Ä—Å–∏–∫", "–í–∞—Å—å–∫–∞", "–ë–∞—Ä—Å–∏–∫",
            "–ú—É—Ä–∑–∏–∫",]
results = Counter(cat_votes)
print(results.most_common(1))
```

## –°–ª–æ–≤–∞—Ä—å —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é

–ë–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è defaultdict –∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è –æ–±—ã—á–Ω—ã—Ö —Å–ª–æ–≤–∞—Ä–µ–π –Ω–∞–º –¥–æ–≤–æ–ª—å–Ω–æ —á–∞—Å—Ç–æ –ø—Ä–∏–¥—ë—Ç—Å—è –ø–∏—Å–∞—Ç—å –æ—Ç–¥–µ–ª—å–Ω—É—é –ª–æ–≥–∏–∫—É –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –Ω–∞–ª–∏—á–∏—è –∫–ª—é—á–∞ –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –Ω–∞—á–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è. –í–æ—Ç —Å–ø–æ—Å–æ–± —ç—Ç–æ–≥–æ –∏–∑–±–µ–∂–∞—Ç—å:

```python
from collections import defaultdict

orders = defaultdict(list)  # –î–ª—è –≤—Å–µ—Ö –Ω–æ–≤—ã—Ö –∫–ª—é—á–µ–π –∑–Ω–∞—á–µ–Ω–∏–µ ‚Äî –Ω–æ–≤—ã–π –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫.
orders['–ú–∞—à–∞'].append('–±—É—Ä–≥–µ—Ä')
orders['–ú–∞—à–∞'].append('—á–∞–π')
orders['–ò–≥–æ—Ä—å'].append('—Å—É—à–∏')
orders
```
