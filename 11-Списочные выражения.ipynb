## –ø—Ä–∏ —Ä–µ—à–µ–Ω–∏–∏ –∑–∞–¥–∞—á —ç—Ç–æ–≥–æ —É—Ä–æ–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–ø–∏—Å–æ—á–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è, –±–µ–∑ –Ω–∏—Ö –∑–∞–¥–∞—á–∞ –Ω–µ –±—É–¥–µ—Ç –ø—Ä–∏–Ω—è—Ç–∞. 

## –¢–∞–∫–∏–µ –¥–µ–ª–∞ `¬Ø\_(„ÉÑ)_/¬Ø`

## –ö—Ä–∞—Ç–∫–æ—Å—Ç—å ‚Äî —Å–µ—Å—Ç—Ä–∞ Python

```python
squares = []
for x in range(10):
    squares.append(x * x)
```

–ò –≤–¥—Ä—É–≥ –Ω–∞—Ç—ã–∫–∞–µ—Ç–µ—Å—å –Ω–∞ —á–µ–π-—Ç–æ –∫–æ–¥:

```python
squares = [x * x for x in range(10)]
```

## –°–ø–∏—Å–æ—á–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è (`list comprehensions`, –≤ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –ø–µ—Ä–µ–≤–æ–¥–∞—Ö –Ω–∞ —Ä—É—Å—Å–∫–∏–π —Ç–∞–∫–∂–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è ‚Äî —Å–ø–∏—Å–æ—á–Ω—ã–µ –≤–∫–ª—é—á–µ–Ω–∏—è) ‚Äî —ç—Ç–æ –Ω–µ –ø—Ä–æ—Å—Ç–æ —Å–∏–Ω—Ç–∞–∫—Å–∏—á–µ—Å–∫–∏–π —Å–∞—Ö–∞—Ä. –≠—Ç–æ –Ω–æ–≤—ã–π —Å–ø–æ—Å–æ–± –¥—É–º–∞—Ç—å. –í—ã –±–æ–ª—å—à–µ –Ω–µ –ø–∏—à–µ—Ç–µ –∫–∞–∫, –∞ –æ–ø–∏—Å—ã–≤–∞–µ—Ç–µ, —á—Ç–æ –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å.

## ‚ò∞ –í—ã—Ä–∞–∂–∞–µ–º—Å—è –≥—Ä–∞–º–æ—Ç–Ω–æ
–ï—â—ë —Ä–∞–∑ –≤–µ—Ä–Ω—ë–º—Å—è –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Å–∏–Ω—Ç–∞–∫—Å–∏—Å.

–ò—Ç–∞–∫, –ø—É—Å—Ç—å –Ω–∞–º –Ω–∞–¥–æ –Ω–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–ª—É—á–∏—Ç —Å–ø–∏—Å–æ–∫ –¥–ª–∏–Ω —Å—Ç—Ä–æ–∫ –∏–∑ —Å–ø–∏—Å–∫–∞ —ç—Ç–∏—Ö —Å–∞–º—ã—Ö —Å—Ç—Ä–æ–∫. –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π –ø–æ–¥—Ö–æ–¥:

1. —Å–¥–µ–ª–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫,
2. –ø—Ä–æ–π–¥—ë–º—Å—è –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º –∏—Å—Ö–æ–¥–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞,
3. –¥–æ–±–∞–≤–∏–º –¥–ª–∏–Ω—É –∫–∞–∂–¥–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –≤ –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫.

```python
excuses = [
    "–ñ–¥–∞–ª –ø–æ–¥—Ö–æ–¥—è—â–µ–≥–æ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫–∞",
    "–ù–µ –±—ã–ª–æ –ø–æ–¥—Ö–æ–¥—è—â–µ–π —Ç–µ—Ç—Ä–∞–¥–∫–∏",
    "–ö–æ—Ç —Å–ø–∞–ª –Ω–∞ –∫–ª–∞–≤–∏–∞—Ç—É—Ä–µ",
    "–°–ª–∏—à–∫–æ–º —Å–æ–ª–Ω–µ—á–Ω–æ –¥–ª—è —Ä–∞–±–æ—Ç—ã",
    "–°–ª–∏—à–∫–æ–º –ø–∞—Å–º—É—Ä–Ω–æ –¥–ª—è –≤–¥–æ—Ö–Ω–æ–≤–µ–Ω–∏—è"
]
result = []
for ex in excuses:
    result.append(len(ex))
print(result)
```

–í —Å–ª—É—á–∞–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å–ø–∏—Å–æ—á–Ω—ã—Ö –≤—ã—Ä–∞–∂–µ–Ω–∏–π –Ω–∞–º –Ω–∞–¥–æ –æ–ø–∏—Å–∞—Ç—å –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ç–æ—Ä—É –∂–µ–ª–∞–µ–º—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:

```python
bed_excuses = [
    "–≥—Ä–∞–≤–∏—Ç–∞—Ü–∏—è —É—Å–∏–ª–∏–ª–∞—Å—å",
    "–æ–¥–µ—è–ª–æ –Ω–µ –æ—Ç–ø—É—Å–∫–∞–ª–æ",
    "—Å–Ω–∞—Ä—É–∂–∏ - —Ä–µ–∞–ª—å–Ω–æ—Å—Ç—å",
    "–º–∏—Ä –ø–æ–∫–∞ –Ω–µ –≥–æ—Ç–æ–≤ –∫ –º–æ–µ–º—É –≤–µ–ª–∏—á–∏—é",
    "—Å—Ç–æ—è–ª –Ω–∞ –∑–∞—Ä—è–¥–∫–µ"
]
result = [len(x) for x in bad_excuses]
```

–ú–æ–∂–Ω–æ —á–∏—Ç–∞—Ç—å —Ç–∞–∫: ¬´–°–æ–∑–¥–∞–π –º–Ω–µ —Å–ø–∏—Å–æ–∫ –∏–∑ –¥–ª–∏–Ω x –¥–ª—è –∫–∞–∂–¥–æ–≥–æ x –∏–∑ —Å–ø–∏—Å–∫–∞ bad_excuses¬ª.

```python
batteries = [90, 50, 20, 100]  # –ü–æ—Å—Ç–∞–≤–∏–ª–∏ –±–∞—Ç–∞—Ä–µ–π–∫–∏ –Ω–∞ –∑–∞—Ä—è–¥–∫—É.
batteries = [min(level + 10, 100) for level in batteries]  # –ó–∞—Ä—è–¥–∏–ª–∏ –±–∞—Ç–∞—Ä–µ–π–∫–∏ –Ω–∞ 10%.
print(batteries)  # [100, 60, 30, 100]
```

–ï—Å–ª–∏ –≤ –∑–∞–¥–∞—á–µ –Ω–∞–ø–∏—Å–∞–Ω–æ —á—Ç–æ-—Ç–æ –ø–æ—Ö–æ–∂–µ–µ –Ω–∞ ¬´–≤–≤–æ–¥–∏—Ç—Å—è —Å–ø–∏—Å–æ–∫ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é¬ª ‚Äî —Å–∞–º–æ–µ –≤—Ä–µ–º—è –¥–æ—Å—Ç–∞–≤–∞—Ç—å —Å–ø–∏—Å–æ—á–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ:

```python
data = [int(x) for x in input().split(',')]
```

–í–∞–∂–Ω—ã–π –º–æ–º–µ–Ω—Ç, —á–∏—Å—Ç–æ –≤–∏–∑—É–∞–ª—å–Ω–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤–æ—Ç —ç—Ç–∏—Ö –¥–≤—É—Ö –∫—É—Å–æ—á–∫–æ–≤ –±—É–¥–µ—Ç –∏–¥–µ–Ω—Ç–∏—á–Ω—ã–º:

```python
print(sum([x ** 2 for x in range(10000000)]))
```

```python
print(sum(x ** 2 for x in range(10000000)))
```

–ù–æ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –≤–æ –≤—Ç–æ—Ä–æ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–∏ –≤–æ–æ–±—â–µ –Ω–∏–∫–∞–∫–æ–≥–æ —Å–ø–∏—Å–∫–∞ –Ω–µ—Ç. –ê —á—Ç–æ –µ—Å—Ç—å? –£–∑–Ω–∞–µ–º —á–µ—Ä–µ–∑ –ø–∞—Ä—É —É—Ä–æ–∫–æ–≤, –∞ –ø–æ–∫–∞ –Ω–µ –∑–∞–±—ã–≤–∞–π—Ç–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ —Å–∫–æ–±–∫–∏.

## ü§î –í—ã–ø–æ–ª–Ω—è–π—Ç–µ –º–æ–∏ —É—Å–ª–æ–≤–∏—è
–ö–æ–≥–¥–∞ –º—ã –≥–æ–≤–æ—Ä–∏–ª–∏ –ø—Ä–æ —É—Å–ª–æ–≤–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã –≤ Python, –º—ã —Å–æ–∑–Ω–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ—à–ª–∏ –º–∏–º–æ —Å–æ–∫—Ä–∞—â—ë–Ω–Ω–æ–π –∑–∞–ø–∏—Å–∏ if ‚Äî —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–≥–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞. –í–æ—Ç —Ç–∞–∫ –≤—ã–≥–ª—è–¥–∏—Ç –µ–≥–æ —Å–∏–Ω—Ç–∞–∫—Å–∏—Å:

`b if condition else c`

–ß–∏—Ç–∞–µ—Ç—Å—è: –≤–µ—Ä–Ω–∏ b, –µ—Å–ª–∏ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —É—Å–ª–æ–≤–∏–µ condition, –∏–Ω–∞—á–µ –≤–µ—Ä–Ω–∏ c. –¢–µ—Ä–Ω–∞—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä, –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç –æ–±—ã—á–Ω–æ–≥–æ if, –Ω–µ –±—ã–≤–∞–µ—Ç –±–µ–∑ else-—á–∞—Å—Ç–∏.

–ü—Ä–∏—á–∏–Ω–∞, –ø–æ –∫–æ—Ç–æ—Ä–æ–π —ç—Ç–æ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä –æ–∫–∞–∑–∞–ª—Å—è —Ç–∞–∫ –¥–∞–ª–µ–∫–æ –æ—Ç —É—Ä–æ–∫–∞ –ø—Ä–æ —É—Å–ª–æ–≤–∏—è, –∫—Ä–æ–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –ø–æ-–Ω–∞—Å—Ç–æ—è—â–µ–º—É –º–æ—â—å —Ç–µ—Ä–Ω–∞—Ä–Ω–æ–≥–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ —Ä–∞—Å–∫—Ä—ã–≤–∞–µ—Ç—Å—è –≤ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ —Å–æ —Å–ø–∏—Å–æ—á–Ω—ã–º–∏ –≤—ã—Ä–∞–∂–µ–Ω–∏—è–º–∏.

–ù–∞–ø—Ä–∏–º–µ—Ä, —É –Ω–∞—Å –µ—Å—Ç—å —Å–ø–∏—Å–æ–∫, —á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞–¥–æ –≤–æ–∑–≤–µ—Å—Ç–∏ –≤ –∫–≤–∞–¥—Ä–∞—Ç, –∞ –Ω–µ—á—ë—Ç–Ω—ã–µ ‚Äî –≤ —Ç—Ä–µ—Ç—å—é —Å—Ç–µ–ø–µ–Ω—å:

```python
numbers = [1, 2, 3, 4, 5]
result = [x ** 2 if x % 2 == 0 else x ** 3 for x in numbers]
print(result)
```

–ú—ã —Ö–æ—Ç–∏–º —Ä–∞–∑–º–µ—Ç–∏—Ç—å –ª—é–¥–µ–π –Ω–∞ –≤–∑—Ä–æ—Å–ª—ã—Ö –∏ –¥–µ—Ç–µ–π –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∏—Ö –≤–æ–∑—Ä–∞—Å—Ç–∞:

```python
ages = [5, 17, 34, 70]
status = ["—Ä–µ–±—ë–Ω–æ–∫" if a < 18 else "–≤–∑—Ä–æ—Å–ª—ã–π" for a in ages]
print(status)
```

## üóëÔ∏è –§–∏–ª—å—Ç—Ä—É–µ–º –≤—ã—Ä–∞–∂–µ–Ω–∏—è
–Ω–æ–≥–¥–∞ –Ω–∞–º –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç–µ—Ä–Ω–∞—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –æ–±—Ä–∞–±–æ—Ç–∞—Ç—å –∏—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫. –ò–Ω–æ–≥–¥–∞ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ, —á—Ç–æ–±—ã –≤ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π —Å–ø–∏—Å–æ–∫ –ø–æ–ø–∞–ª–∏ –Ω–µ –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã, –∞ —Ç–æ–ª—å–∫–æ —Ç–µ, –∫–æ—Ç–æ—Ä—ã–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—Ç –∫–∞–∫–æ–º—É-–ª–∏–±–æ —É—Å–ª–æ–≤–∏—é.

```python
evens = []
for x in range(10): 
    if x % 2 == 0:
        evens.append(x ** 2)
print(evens)
```

```python
evens = [x ** 2 for x in range(10) if x % 2 == 0]
print(evens)
```

–º–∏–≥—Ä–∞—Ü–∏—è –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤ –∏–∑ –ø—Ä–∏–≤—ã—á–Ω–æ–≥–æ –∫–æ–¥–∞ –≤ —Å–ø–∏—Å–æ—á–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –≤—ã–≥–ª—è–¥–∏—Ç –≤–æ—Ç —Ç–∞–∫:
<img src="img/11-01.jpg" width=900 height=400 />

```python
foods = ["—è–±–ª–æ–∫–æ", "–∫–∞–º–µ–Ω—å", "–≥—Ä—É—à–∞", "–≥–≤–æ–∑–¥—å", "–±–∞–Ω–∞–Ω", "—è–±–ª–æ–∫–æ"]
edible = [food for food in foods if food in ("—è–±–ª–æ–∫–æ", "–≥—Ä—É—à–∞", "–±–∞–Ω–∞–Ω")]
print(edible)
```

```python
words = ["—è", "–ª—é–±–ª—é", "Python", "–∏", "–∫–æ—Ñ–µ"]
result = [word.upper() for word in words if len(word) > 2]
print(result)
```

## ü§Ø –ó–∞–Ω—è—Ç–Ω–∞—è –∫–æ–º–±–∏–Ω–∞—Ü–∏—è
–í —Å–ø–∏—Å–æ—á–Ω–æ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–∏ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ü–∏–∫–ª–æ–≤. –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –≤–Ω–µ—à–Ω–∏–º —Å—á–∏—Ç–∞–µ—Ç—Å—è —Ç–æ—Ç, –∫–æ—Ç–æ—Ä—ã–π –Ω–∞–ø–∏—Å–∞–Ω —Ä–∞–Ω—å—à–µ.

–î–æ–ø—É—Å—Ç–∏–º –º—ã –∑–∞—Ö–æ—Ç–µ–ª–∏ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –≤—Å–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –ø—Ä–æ–¥—É–∫—Ç–æ–≤ –≤ –Ω–∞—à–µ–º —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–µ:

```python
ingredients = ["—Ö–ª–µ–±", "–∫–æ–ª–±–∞—Å–∞", "—Å—ã—Ä", "–≤–∞—Ä–µ–Ω—å–µ"]
combinations = [(a, b) for a in ingredients for b in ingredients if a != b]
print(combinations)
# –ï—â—ë —Ä–∞–∑: –Ω–µ –∑–∞–±—ã–≤–∞–µ–º –ø—Ä–æ –ø–æ—Ä—è–¥–æ–∫ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ü–∏–∫–ª–æ–≤ ‚Äî —Å–Ω–∞—á–∞–ª–∞ –≤–Ω–µ—à–Ω–∏–π, –ø–æ—Ç–æ–º –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π!
```

–í—Å–µ –¥–∞—Ç—ã –≥–æ–¥–∞, –∫–æ–≥–¥–∞ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç –ø–∏–ª –∫–æ—Ñ–µ (–¥–∞–∂–µ –Ω–µ—Å—É—â–µ—Å—Ç–≤—É—é—â–∏–µ).

```python
coffee_days = [(day, month) for month in range(1, 13) for day in range(1, 32)]
print(coffee_days)
```

## üåê –í–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ –≤ –º–∞—Ç—Ä–∏—Ü—É
–°–ø–∏—Å–æ—á–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –º–∞—Ç—Ä–∏—Ü, –¥–ª—è —ç—Ç–æ–≥–æ –Ω—É–∂–Ω–æ –≤—Å–µ–≥–æ –ª–∏—à—å –∑–∞—Å—É–Ω—É—Ç—å –æ–¥–Ω–æ —Å–ø–∏—Å–æ—á–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ (–∫–æ—Ç–æ—Ä–æ–µ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —ç–ª–µ–º–µ–Ω—Ç—ã —Å—Ç—Ä–æ–∫–∏) –≤ –¥—Ä—É–≥–æ–µ (–∫–æ—Ç–æ—Ä–æ–µ, —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ, –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç —Å—Ç—Ä–æ–∫–∏ –≤ –º–∞—Ç—Ä–∏—Ü—É).

```python
matrix = [[x * y for y in range(1, 5)] for x in range(1, 4)]
print(matrix)
```

–ù–∞ –∫–∞–∂–¥—É—é –∏—Ç–µ—Ä–∞—Ü–∏—é —Ü–∏–∫–ª–∞ –≤–æ –≤–Ω–µ—à–Ω–µ–º —Å–ø–∏—Å–æ—á–Ω–æ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–∏ –ø–æ–ª–Ω–æ—Å—Ç—å—é –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–µ —Å–ø–∏—Å–æ—á–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ.

–î–∞–≤–∞–π—Ç–µ —Å–¥–µ–ª–∞–µ–º —Å –ø–æ–º–æ—â—å—é –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö —Å–ø–∏—Å–æ—á–Ω—ã—Ö –≤—ã—Ä–∞–∂–µ–Ω–∏–π ¬´—à–∞—Ö–º–∞—Ç–Ω—É—é –¥–æ—Å–∫—É¬ª:

```python
board = [['‚¨õ' if (i + j) % 2 else '‚¨ú' for j in range(8)] for i in range(8)]
print(*board, sep='\n')
```

–° –≥–µ–Ω–µ—Ä–∞—Ü–∏–µ–π –º–∞—Ç—Ä–∏—Ü –µ—Å—Ç—å –Ω–µ–±–æ–ª—å—à–æ–π –ø–æ–¥–≤–æ—Ö. –°–∫–∞–∂–∏—Ç–µ, –≤–µ–¥—å –ø—Ä–∞–≤–¥–∞ –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Å–ø–∏—Å–∫–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Å–ø–∏—Å–∫–æ–≤ —Ö–æ—á–µ—Ç—Å—è –Ω–∞–ø–∏—Å–∞—Ç—å —á—Ç–æ-—Ç–æ –≤—Ä–æ–¥–µ:

```python
row = [0] * 3
grid = [row] * 3
print(grid)
```

–ù–∞ –ø–µ—Ä–≤—ã–π –≤–∑–≥–ª—è–¥ –≤—Å—ë —Ö–æ—Ä–æ—à–æ, –Ω–æ –ø—Ä–æ–±–ª–µ–º–∞ –ø–æ–∫–∞–∂–µ—Ç —Å–µ–±—è —Å—Ä–∞–∑—É –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–≥–æ –∂–µ –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞:

```python
grid[0][0] = 1
print(grid)
```

–ß—Ç–æ–±—ã —ç—Ç–æ–≥–æ –∏–∑–±–µ–∂–∞—Ç—å, –Ω–∞–¥–æ —è–≤–Ω–æ –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ —Å–æ–∑–¥–∞–≤–∞—Ç—å –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å–ø–∏—Å–æ–∫:

```python
grid = [[0] * 3 for _ in range(3)]
grid[0][0] = 1
print(grid)
```

## üòµ‚Äçüí´ –°–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ —Ö–æ—Ä–æ—à–µ–≥–æ ‚Äî —ç—Ç–æ —Ç–æ–∂–µ –ø–ª–æ—Ö–æ
–ò –≤–æ–æ–±—â–µ, python-—Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫–∏ –ª—é–±—è—Ç –≤ –∫–∞—á–µ—Å—Ç–≤–µ —É–ø—Ä–∞–∂–Ω–µ–Ω–∏—è –ø—ã—Ç–∞—Ç—å—Å—è —Ä–µ—à–∏—Ç—å –ª—é–±—É—é –∑–∞–¥–∞—á—É —Å –ø–æ–º–æ—â—å—é one-liner (–ø—Ä–æ–≥—Ä–∞–º–º—ã –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É), –Ω–æ –≤ —Ä–µ–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–µ–∫—Ç–∞—Ö —Å–ø–∏—Å–æ—á–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –¥–ª–∏–Ω–æ–π –≤ —Ç—ã—Å—è—á–∏ —Å–∏–º–≤–æ–ª–æ–≤ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ –≤—Ä–µ–¥—è—Ç, –∞ –Ω–µ –¥–µ–ª–∞—é—Ç –∫–æ–¥ –ª—É—á—à–µ.

–ï—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Å–ø–∏—Å–æ—á–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç –ø—Ä–æ–≥—Ä–∞–º–º—É –≤ —Ä–µ–±—É—Å, —Ç–æ –ª—É—á—à–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –æ—Ç–¥–µ–ª—å–Ω—ã–π —Ü–∏–∫–ª:

```python
# –ü–ª–æ—Ö–æ:
result = [x if x % 2 == 0 else (x ** 2 if x > 5 else x / 2) for x in range(10)]

# –•–æ—Ä–æ—à–æ:
result = []
for x in range(10):
    if x % 2 == 0:
        result.append(x)
    else:
        result.append(x ** 2 if x > 5 else x / 2)
```

## –ï—Å–ª–∏ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –≤ 3 –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö —É—Ä–æ–≤–Ω—è ‚Äî –ø–æ—Ä–∞ –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—å—Å—è –∏ –ø–æ–¥—ã—à–∞—Ç—å.

```python
# –ü–ª–æ—Ö–æ
[print(f"–ó–Ω–∞—á–µ–Ω–∏–µ: {x}") for x in data]

# –õ—É—á—à–µ
for x in data:
    print(f"–ó–Ω–∞—á–µ–Ω–∏–µ: {x}")
```

## ü§® –ß—Ç–æ —Ç—ã —Ç–∞–∫–æ–µ?
–¢–∞–∫ –∫–∞–∫ Python ‚Äî —è–∑—ã–∫ —Å –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–π —Ç–∏–ø–∏–∑–∞—Ü–∏–µ–π, –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ç–æ—Ä—É –≤ –∫–∞–∂–¥—ã–π –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∑–Ω–∞—Ç—å, –∫–∞–∫–æ–≥–æ —Ç–∏–ø–∞ –∑–Ω–∞—á–µ–Ω–∏–µ —Ö—Ä–∞–Ω–∏—Ç—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π. –î–ª—è —ç—Ç–æ–≥–æ, –∫–∞–∂–¥—ã–π –æ–±—ä–µ–∫—Ç –≤ Python, –ø–æ–º–∏–º–æ –∑–Ω–∞—á–µ–Ω–∏—è, —Ö—Ä–∞–Ω–∏—Ç –µ—â—ë –∏ —Ç–∏–ø. –≠—Ç–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –¥–æ—Å—Ç—É–ø–Ω–∞ –∏ –Ω–∞–º –∫–∞–∫ —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫–∞–º.

–ú—ã –≤—Å–µ–≥–¥–∞ –º–æ–∂–µ–º –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, —Ö—Ä–∞–Ω–∏—Ç –ª–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –æ–±—ä–µ–∫—Ç –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–≥–æ —Ç–∏–ø–∞ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ `isinstance()`.

```python
x = [n for n in range(3)]
print(isinstance(x, list))
```

```python
print(isinstance({}, set))
```

```python
print(isinstance(3, int))
```

```python
mixed = [42, "—ç—Å–ø—Ä–µ—Å—Å–æ", 3.14, "–∫–∞–ø—É—á–∏–Ω–æ", ["—Å–ø–∏—Å–æ–∫"]]
strings = [item for item in mixed if isinstance(item, str)]
print(strings)
```
