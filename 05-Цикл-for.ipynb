#### –ú—ã –≥–æ–≤–æ—Ä–∏–ª–∏ (–∏ –ø—Ä–æ–¥–æ–ª–∂–∞–µ–º –≥–æ–≤–æ—Ä–∏—Ç—å), —á—Ç–æ –Ω–∞–¥–æ –¥–∞–≤–∞—Ç—å –æ—Å–º—ã—Å–ª–µ–Ω–Ω—ã–µ –∏–º–µ–Ω–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º. –ù–æ, –ø–æ —Å–æ–≥–ª–∞—à–µ–Ω–∏—é PEP8, –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª–∞, –∫–æ—Ç–æ—Ä—ã–µ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –∫–∞–∫ —Å—á—ë—Ç—á–∏–∫–∏, –º–æ–∂–Ω–æ –Ω–∞–∑—ã–≤–∞—Ç—å –æ–¥–Ω–æ–π –±—É–∫–≤–æ–π ‚Äî `i`, `j`, `k`. –ú–æ–∂–Ω–æ –≤–æ–æ–±—â–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –∏–º–µ–Ω–∏ –Ω–∏–∂–Ω–µ–µ –ø–æ–¥—á—ë—Ä–∫–∏–≤–∞–Ω–∏–µ `_`, –Ω–æ —Ç–æ–ª—å–∫–æ –≤ —Ç–æ–º —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ –º—ã —Ö–æ—Ç–∏–º —Å–¥–µ–ª–∞—Ç—å –∞–∫—Ü–µ–Ω—Ç: –≤–Ω—É—Ç—Ä–∏ —Ü–∏–∫–ª–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –Ω–∏–∫–∞–∫ –Ω–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∏ –Ω—É–∂–Ω–∞ –ª–∏—à—å –¥–ª—è –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏–∏ –Ω—É–∂–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π.

### –ó–∞–¥–∞—á–∞ ¬´–†–µ—à–∞—Ç—å –∑–∞–¥–∞—á–∏¬ª

–î–∞–≤–∞–π—Ç–µ –≤–Ω–∏–º–∞—Ç–µ–ª—å–Ω–µ–µ –ø–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–æ–∑–¥–∞—ë—Ç range().

```python
for i in range(5):  
    print(f"–î–µ–Ω—å {i}: –Ø –≤—Å—ë –µ—â—ë –Ω–µ —É–º–µ—é –ª–µ—Ç–∞—Ç—å üê¶")
```

–•–º, –∞ —á—Ç–æ, –µ—Å–ª–∏ –º—ã –ø–∏—à–µ–º –ø—Ä–æ–≥—Ä–∞–º–º—ã –Ω–µ –¥–ª—è –¥—Ä—É–≥–∏—Ö –ø—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç–æ–≤, –∞ –¥–ª—è –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö –ª—é–¥–µ–π, –∫–æ—Ç–æ—Ä—ã–µ —Å—á–∏—Ç–∞—é—Ç —Å –µ–¥–∏–Ω–∏—Ü—ã? –ï—Å—Ç—å –¥–≤–∞ —Å–ø–æ—Å–æ–±–∞ —Ä–µ—à–µ–Ω–∏—è –ø—Ä–æ–±–ª–µ–º—ã:

```python
print("–ü–æ—Å—á–∏—Ç–∞—é –æ—Ç –µ–¥–∏–Ω–∏—Ü—ã –¥–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ)")
for i in range(int(input())):
    print(i + 1)
```

–í—ã–∑–≤–∞—Ç—å `range()` –¥–≤—É–º—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ ‚Äî –Ω–∞—á–∞–ª–æ–º –∏ –∫–æ–Ω—Ü–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–∞:

```python
print("–ü–æ—Å—á–∏—Ç–∞—é –æ—Ç –µ–¥–∏–Ω–∏—Ü—ã –¥–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ)")
for i in range(1, int(input()) + 1):  # –ü—Ä–∏–±–∞–≤–∏–ª–∏ –µ–¥–∏–Ω–∏—Ü—É, —á—Ç–æ–±—ã –≤–æ—à–ª–æ —Å–∞–º–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–µ —á–∏—Å–ª–æ.
    print(i)
```

–ó–Ω–∞—á–∏—Ç, `range()` —É–º–µ–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –Ω–∞—Å—Ç—Ä–æ–π–∫—É –Ω–µ —Ç–æ–ª—å–∫–æ –∫–æ–Ω—Ü–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ (–Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ), –Ω–æ –∏ –Ω–∞—á–∞–ª–∞ (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ). –ê –µ—â—ë –æ–Ω –º–æ–∂–µ—Ç –≤—ã–¥–∞–≤–∞—Ç—å —á–∏—Å–ª–∞ –Ω–µ –ø–æ–¥—Ä—è–¥, –∞ —Å –Ω–µ–∫–æ—Ç–æ—Ä—ã–º —à–∞–≥–æ–º. –î–æ–ø—É—Å—Ç–∏–º, –Ω–∞–º –Ω–∞–¥–æ –≤—ã–≤–µ—Å—Ç–∏ –≤—Å–µ —á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–≥–æ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. –ú—ã –º–æ–∂–µ–º –Ω–∞–ø–∏—Å–∞—Ç—å `if` –≤–Ω—É—Ç—Ä–∏ —Ü–∏–∫–ª–∞:

```python
for i in range(1, int(input()) + 1):
    if i % 2 == 0:  # –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤—Å–µ –∏ –∏—â–µ–º –ø–æ–¥—Ö–æ–¥—è—â–µ–µ.
        print(f'{i} - —á—ë—Ç–Ω–æ–µ')
```

–ê –º–æ–∂–µ–º –∑–∞–¥–∞—Ç—å —à–∞–≥ –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Ç—Ä–µ—Ç—å–µ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ range():

```python
for i in range(2, int(input()) + 1, 2):  # –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –∏–¥—ë–º —Å –¥–≤–æ–π–∫–∏, –∞ —Ç–æ –ø–µ—Ä–µ–±—Ä–∞–ª–∏ –±—ã –≤—Å–µ –Ω–µ—á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞. 
    print(f'{i} - —á—ë—Ç–Ω–æ–µ')  # –í—ã–≤–æ–¥–∏–º —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω–æ–µ.
```

### –ü–æ–¥—ã—Ç–æ–∂–∏–º. `range()` –º–æ–∂–Ω–æ –≤—ã–∑—ã–≤–∞—Ç—å —Å —Ä–∞–∑–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤:

* –° –æ–¥–Ω–∏–º. `range(stop)` —Å–æ–∑–¥–∞—Å—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª –æ—Ç `–Ω—É–ª—è` –¥–æ `stop` –Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.
* –° –¥–≤—É–º—è. `range(start, stop)` ‚Äî –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ—Ç `start` –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –¥–æ `stop` –Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.
* –° —Ç—Ä–µ–º—è. `range(start, stop, step)` ‚Äî –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ—Ç `start` –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –¥–æ `stop` –Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ —Å —à–∞–≥–æ–º `step`.

### –ó–∞–¥–∞—á–∞ ¬´–ê —Ç–µ–ø–µ—Ä—å –Ω—É–º–µ—Ä–∞—Ü–∏—è!¬ª

```python
name = input()
for i in range(int(input()), int(input()) + 1):
    print(f"{i}-–π {name}.")
```

–ù–∏–∫—Ç–æ –Ω–µ –∑–∞–ø—Ä–µ—â–∞–µ—Ç –Ω–∞–º —Å–¥–µ–ª–∞—Ç—å —à–∞–≥ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º. –¢–æ–≥–¥–∞, —á—Ç–æ–±—ã —Ü–∏–∫–ª –∑–∞–ø—É—Å—Ç–∏–ª—Å—è —Ö–æ—Ç—å —Ä–∞–∑, –Ω–∞–¥–æ –∑–∞–¥–∞—Ç—å start –±–æ–ª—å—à–∏–º –∏–ª–∏ —Ä–∞–≤–Ω—ã–º stop:

```python
# –°—á–∏—Ç–∞–µ–º –æ—Ç 10 –¥–æ 0 (–Ω–µ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ) —Å —à–∞–≥–æ–º -1.  
for countdown in range(10, 0, -1):  
    print(f"üöÄ {countdown}...")  
print("–¢—É –∑–µ –º—É–Ω!")
```

–ö–æ–≥–¥–∞ –º—ã –≥–æ–≤–æ—Ä–∏–º –ø—Ä–æ `range()`, —Ç–æ —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ –∏–∑–±–µ–≥–∞–µ–º —Ñ—Ä–∞–∑—ã ¬´–≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å¬ª, –ø–æ—Ç–æ–º—É —á—Ç–æ —á–∞—Å—Ç–æ `range()` –Ω–∞–∑—ã–≤–∞—é—Ç –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–æ–º, –∞ —ç—Ç–æ –Ω–µ —Ç–∞–∫. –ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —ç—Ç–æ (—Å–µ–π—á–∞—Å –±—É–¥–µ—Ç —Å—Ç—Ä–∞—à–Ω–æ) ‚Äî –ª–µ–Ω–∏–≤–∞—è –Ω–µ–∏–∑–º–µ–Ω—è–µ–º–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–π –º–æ–∂–Ω–æ —Å–æ–∑–¥–∞–≤–∞—Ç—å –∏—Ç–µ—Ä–∞—Ç–æ—Ä—ã. (–ê-–∞-–∞-–∞-–∞-–∞! üò±) –ö–æ–≥–¥–∞ –º—ã –¥–æ–±–µ—Ä—ë–º—Å—è –¥–æ –ø–æ–¥—Ä–æ–±–Ω–æ—Å—Ç–µ–π, —ç—Ç–∞ —Ñ—Ä–∞–∑–∞ —É–∂–µ –Ω–µ –±—É–¥–µ—Ç –∑–≤—É—á–∞—Ç—å —Ç–∞–∫ –ø—É–≥–∞—é—â–µ, –æ–±–µ—â–∞–µ–º. –ê –ø–æ–∫–∞ –≤–∞–∂–Ω–æ –∑–Ω–∞—Ç—å, —á—Ç–æ `range()` –Ω–µ —Å–æ–∑–¥–∞—ë—Ç –≤—Å—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª —Å—Ä–∞–∑—É, –∞ –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–æ–ª—É—á–∞—Ç—å –ø–æ –æ–¥–Ω–æ–º—É —á–∏—Å–ª—É –∑–∞ —Ä–∞–∑. –ü–æ—ç—Ç–æ–º—É `range(1_000_000_000_000_000)` –Ω–µ —Å—ä–µ—Å—Ç –≤—Å—é –æ–ø–µ—Ä–∞—Ç–∏–≤–Ω—É—é –ø–∞–º—è—Ç—å –≤–∞—à–µ–≥–æ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–∞.

```python
my_range = range(1, 100, 2)  # –ú–æ–∂–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é.
```

```python
for i in my_range:
    print(i)
```

## üèéÔ∏èüí® –¢—é–Ω–∏–Ω–≥—É–µ–º print

–§—É–Ω–∫—Ü–∏—è `print()` –≤ Python –æ—á–µ–Ω—å –≥–∏–±–∫–∞—è. –£ –Ω–µ—ë –µ—Å—Ç—å –∏–º–µ–Ω–æ–≤–∞–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã (keyword arguments), –ø–æ–∑–≤–æ–ª—è—é—â–∏–µ –Ω–∞—Å—Ç—Ä–æ–∏—Ç—å —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å –º–µ–∂–¥—É –≤—ã–≤–æ–¥–∏–º—ã–º–∏ —á–∞—Å—Ç—è–º–∏ –∏ —Ç–µ–º, —á—Ç–æ –±—É–¥–µ—Ç –≤ –∫–æ–Ω—Ü–µ –≤—ã–≤–æ–¥–∞:

### sep (separator)
—Å—Ç—Ä–æ–∫–∞-—Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å, –∫–æ—Ç–æ—Ä–∞—è –≤—Å—Ç–∞–≤–ª—è–µ—Ç—Å—è –º–µ–∂–¥—É –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏ –ø—Ä–∏ –ø–µ—á–∞—Ç–∏. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é sep=" " (–ø—Ä–æ–±–µ–ª). –ú–æ–∂–Ω–æ —Å—Ç–∞–≤–∏—Ç—å –ª—é–±–æ–π —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—å: –∑–∞–ø—è—Ç—É—é, —Ç–∏—Ä–µ, —Å–º–∞–π–ª, —Å–µ–∫—Ä–µ—Ç–Ω–æ–µ –ø–æ—Å–ª–∞–Ω–∏–µ –Ω–∞ 10 –ª–∏—Å—Ç–∞—Ö ‚Äî —á—Ç–æ —É–≥–æ–¥–Ω–æ, –ª–∏—à—å –±—ã —ç—Ç–æ –±—ã–ª–∞ —Å—Ç—Ä–æ–∫–∞.

```python
print("Cats", "Dogs", "Pythons", sep=" vs ")
# –í—ã–≤–æ–¥: Cats vs Dogs vs Pythons
```

### end
—Å—Ç—Ä–æ–∫–∞, –∫–æ—Ç–æ—Ä–∞—è –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –≤ –∫–æ–Ω—Ü–µ –≤—ã–≤–æ–¥–∞. –ü–æ —É–º–æ–ª—á–∞–Ω–∏—é `end="\n"`, —Ç–æ –µ—Å—Ç—å –∫–∞–∂–¥—ã–π `print()` –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –ø–µ—Ä–µ–≤–æ–¥–æ–º —Å—Ç—Ä–æ–∫–∏. –°–¥–µ–ª–∞–µ–º —Ç–∞–∫, —á—Ç–æ–±—ã –¥–≤–∞ –≤—ã–∑–æ–≤–∞ print –Ω–∞–ø–µ—á–∞—Ç–∞–ª–∏ –≤ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ–∫–µ:

```python
print("–ü—Ä–∏–≤–µ—Ç", end=", ")
print("–ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å", end="!\n")
# –í—ã–≤–æ–¥: –ü—Ä–∏–≤–µ—Ç, –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å! + –ø–µ—Ä–µ–≤–æ–¥ –Ω–∞ —Å–ª–µ–¥—É—é—â—É—é —Å—Ç—Ä–æ–∫—É.
```

### –° —Ä—É—Å—Å–∫–∏–º –∞–ª—Ñ–∞–≤–∏—Ç–æ–º –µ—Å—Ç—å –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ —É –±—É–∫–≤ `–Å` –∏ `—ë`.

–ê —Ç–∞–∫ –º–æ–∂–Ω–æ —Ä–∞—Å—à–∏—Ñ—Ä–æ–≤–∞—Ç—å –ø–æ—Å–ª–∞–Ω–∏–µ –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏—è –ø—É–Ω–∫—Ç–∞ —É—á–µ–±–Ω–∏–∫–∞:

–° –ø–æ–º–æ—â—å—é –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π –≤ –∫–æ–¥ —Å–∏–º–≤–æ–ª–∞ –∏ –æ–±—Ä–∞—Ç–Ω–æ —Ä–µ—à–∞—é—Ç—Å—è –∑–∞–¥–∞—á–∏ –Ω–∞ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é —à–∏—Ñ—Ä–æ–≤ –∑–∞–º–µ–Ω—ã, –Ω–∞–ø—Ä–∏–º–µ—Ä —à–∏—Ñ—Ä–∞ –¶–µ–∑–∞—Ä—è: –ø—Ä–∏ —à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–∏ –∫–∞–∂–¥–∞—è –±—É–∫–≤–∞ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ —Å–æ–æ–±—â–µ–Ω–∏—è —Å–¥–≤–∏–≥–∞–µ—Ç—Å—è –Ω–∞ —Å–∫–æ–ª—å–∫–æ-—Ç–æ –ø–æ–∑–∏—Ü–∏–π –≤ –æ–¥–Ω—É —Å—Ç–æ—Ä–æ–Ω—É, –∞ –ø—Ä–∏ —Ä–∞—Å—à–∏—Ñ—Ä–æ–≤–∫–µ —Å–º–µ—â–µ–Ω–∏–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –≤ –æ–±—Ä–∞—Ç–Ω—É—é —Å—Ç–æ—Ä–æ–Ω—É:

```python
secret = ''
shift = int(input())
for char in 'HELLO':
    secret += chr(ord(char) + shift)
print(secret)  # –†–∞—Å—à–∏—Ñ—Ä–æ–≤–∫–∞: —Å–º–µ—â–µ–Ω–∏–µ –Ω–∞–∑–∞–¥ –Ω–∞ -shift.
```
