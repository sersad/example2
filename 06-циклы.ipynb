```python
count = 1
while count < 100:
    if count % 5 == 0:
        continue  # –ú—ã –Ω–µ –ø–æ–º–µ–Ω—è–ª–∏ —Å—á—ë—Ç—á–∏–∫ –∏ –æ–±—Ä–µ—á–µ–Ω—ã –Ω–∞ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ–µ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–µ –∏—Ç–µ—Ä–∞—Ü–∏–∏ —Å count == 5.
    print(count)
    count += 1
```

–ï—Å–ª–∏ –≤–Ω–µ—à–Ω–∏–π —Ü–∏–∫–ª –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è 1000 —Ä–∞–∑, –∞ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π ‚Äî —Ç–æ–∂–µ 1000, –≤—Å–µ–≥–æ –∏—Ç–µ—Ä–∞—Ü–∏–π –±—É–¥–µ—Ç 1_000_000. –≠—Ç–æ –µ—Å–ª–∏ –≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª –æ–¥–∏–Ω. –ï—Å–ª–∏ –µ—Å—Ç—å –≤–ª–æ–∂–µ–Ω–Ω–æ—Å—Ç—å —Ç—Ä–µ—Ç—å–µ–≥–æ —É—Ä–æ–≤–Ω—è, —Ç–æ –±—É–¥–µ—Ç —É–∂–µ —Ç—Ä–µ—Ç—å—è —Å—Ç–µ–ø–µ–Ω—å –∏ 1_000_000_000, –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ. –≠—Ç–æ –∫–∞–∫ –ø—ã—Ç–∞—Ç—å—Å—è –ø–µ—Ä–µ—Å—á–∏—Ç–∞—Ç—å –≤—Å–µ –ø–µ—Å—á–∏–Ω–∫–∏ –Ω–∞ –ø–ª—è–∂–µ ‚Äî —Ç–æ –µ—Å—Ç—å –º–µ–¥–ª–µ–Ω–Ω–æ

–î–∞, –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –≤—Å–µ–≥–¥–∞ –≤ —É—á–µ–±–Ω—ã—Ö –∑–∞–¥–∞—á–∞—Ö —ç—Ç–æ ¬´–º–µ–¥–ª–µ–Ω–Ω–æ, –Ω–æ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ¬ª, —Ö–æ—Ç—è –∏–Ω–æ–≥–¥–∞ (–Ω–µ –≤ —ç—Ç–æ–º —É—Ä–æ–∫–µ) –≤–∞–º –±—É–¥—É—Ç –≤—Å—Ç—Ä–µ—á–∞—Ç—å—Å—è –∑–∞–¥–∞—á–∏, –≥–¥–µ –Ω–∞–¥–æ –±—É–¥–µ—Ç –ø—Ä–∏–¥—É–º–∞—Ç—å –±–æ–ª–µ–µ —Ö–∏—Ç—Ä—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö —Ü–∏–∫–ª–æ–≤, –∏–Ω–∞—á–µ –ø—Ä–æ–≥—Ä–∞–º–º–∞ –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è –Ω–µ–ø–æ–∑–≤–æ–ª–∏—Ç–µ–ª—å–Ω–æ –¥–æ–ª–≥–æ.

```python
k = int(input())
for i in range(1, 10):
    print(f"{i} * {k} = {k * i}", end='\t')
```

```python
k = int(input())
for j in range(1, k + 1):
    for i in range(1, 10):
        print(f"{i} * {j} = {j * i}", end='\t')
    print()  # –ü—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ –ø–æ—Å–ª–µ –≤—Å–µ—Ö –∏—Ç–µ—Ä–∞—Ü–∏–π –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–≥–æ —Ü–∏–∫–ª–∞.
```

 ### –ü–æ–∫–∞ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π —Ü–∏–∫–ª –Ω–µ –æ—Ç—Ä–∞–±–æ—Ç–∞–µ—Ç –≤—Å–µ –∏—Ç–µ—Ä–∞—Ü–∏–∏ –ø—Ä–∏ —Ç–µ–∫—É—â–µ–º j, –≤–Ω–µ—à–Ω–∏–π —Ü–∏–∫–ª –Ω–µ —Å–¥–≤–∏–Ω–µ—Ç—Å—è.

### –ó–∞–¥–∞—á–∞ ¬´–ë—ã—Ç—å –∏–ª–∏¬ª

### –ó–∞–¥–∞—á–∞ ¬´–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–Ω—ã–π —Ä–µ–∫–æ—Ä–¥¬ª

## ‚úàÔ∏è –ö–∞—Ç–∞–ø—É–ª—å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∏–∑ –ª—é–±–æ–π –≥–ª—É–±–∏–Ω—ã
–û–ø–µ—Ä–∞—Ç–æ—Ä—ã `break` –∏ `continue` –∫–ª–∞—Å—Å–Ω—ã–µ, –Ω–æ –æ–Ω–∏ —Ä–∞–±–æ—Ç–∞—é—Ç —Ç–æ–ª—å–∫–æ –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ –≤ —Ç–æ–º —Ü–∏–∫–ª–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞—Ö–æ–¥—è—Ç—Å—è. –ù–æ —á—Ç–æ –¥–µ–ª–∞—Ç—å, –µ—Å–ª–∏ –Ω–∞–º –Ω–∞–¥–æ –≤—ã—Ä–≤–∞—Ç—å—Å—è –∏–∑ –≤—Ç–æ—Ä–æ–≥–æ —É—Ä–æ–≤–Ω—è –≤–ª–æ–∂–µ–Ω–Ω–æ—Å—Ç–∏? –ï—Å—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–ø–æ—Å–æ–±–æ–≤.

```python
# –û–¥–∏–Ω break —Ö–æ—Ä–æ—à–æ, –∞ –≤ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ —Å —Ñ–ª–∞–≥–æ–º ‚Äî –µ—â—ë –ª—É—á—à–µ:
print("–ù–∞—à–∞ –ø—Ä–æ–≥—Ä–∞–º–º–∞ –±–æ–∏—Ç—Å—è —á–∏—Å–µ–ª –±–æ–ª—å—à–µ 700.")
runaway = False
for j in range(1, 100):
    for i in range(1, 100):
        if i * j > 700:
            print('–ê-–∞-–∞-–∞-–∞-–∞, –ø–æ–ø–∞–ª–æ—Å—å —Å–ª–∏—à–∫–æ–º –±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ!')
            runaway = True  # –°—Ç–∞–≤–∏–º —Ñ–ª–∞–≥.
            break  # –í—ã—Ö–æ–¥–∏–º –∏–∑ –≤–ª–æ–∂–µ–Ω–Ω–æ–≥–æ —Ü–∏–∫–ª–∞.
        print(i ** j)
    if runaway:
        break  # –í—ã—Ö–æ–¥–∏–º –∏–∑ –≤–Ω–µ—à–Ω–µ–≥–æ —Ü–∏–∫–ª–∞.
```

### –ó–∞–¥–∞—á–∞ ¬´–°—É–º–º–∞ –Ω–µ—á—ë—Ç–Ω—ã—Ö¬ª

### üîé –ö–∞–∫ –∏—Å–∫–∞—Ç—å —á—ë—Ä–Ω—É—é –∫–æ—à–∫—É –≤ —Ç—ë–º–Ω–æ–π –∫–æ–º–Ω–∞—Ç–µ, –∞ –ø–æ—Ç–æ–º –ø–æ–Ω—è—Ç—å, —á—Ç–æ —ç—Ç–æ –±—ã–ª –µ–Ω–æ—Ç
  –°–∞–º—ã–π –±–∞–∑–æ–≤—ã–π –ø—É—Ç—å ¬´–æ—Ç–ª–∞–¥–∫–∏¬ª ‚Äî –Ω–∞–≤—Ç—ã–∫–∞—Ç—å print() –≤ —Å—Ç—Ä–∞—Ç–µ–≥–∏—á–µ—Å–∫–∏–µ –º–µ—Å—Ç–∞, 

```python
print('–¢—Ä–µ–Ω–∞–∂—ë—Ä –ø–æ –≤–≤–æ–¥—É –ø–∞–ª–∏–Ω–¥—Ä–æ–º–∞:')
while True:
    print('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ-–ø–∞–ª–∏–Ω–¥—Ä–æ–º:')
    number = n = int(input())
    reverse = 0
    while n > 0:
        reverse = reverse * 10 + n % 10
        n //= 10
        print(f"n: {n}, reverse: {reverse} ‚Üí –ú–æ–∂–µ—Ç, —É–∂–µ —Å–¥–∞–¥–∏—Ç–µ—Å—å, –ø–∞–ª–∏–Ω–¥—Ä–æ–º—ã?")  # –®–ø–∏–æ–Ω–∏–º!
    if number == reverse:
        print('–í—ã –≤–≤–µ–ª–∏ –ø–∞–ª–∏–Ω–¥—Ä–æ–º! –ü—Ä–æ–≥—Ä–∞–º–º–∞ –æ—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞.')
        break
    print('–í–≤–µ–¥—ë–Ω–Ω–æ–µ —á–∏—Å–ª–æ –Ω–µ –ø–∞–ª–∏–Ω–¥—Ä–æ–º, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â—ë —Ä–∞–∑.')
```
