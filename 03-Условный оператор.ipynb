–ò `if`, –∏ `else` –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –±–ª–æ—á–Ω—ã–º –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞–º, –¥–ª—è –ª—é–±–æ–≥–æ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –≤ Python –≤–µ—Ä–Ω–æ –ø—Ä–∞–≤–∏–ª–æ ‚Äî —Å—Ç—Ä–æ—á–∫–∞ —Å –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–º –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –¥–≤–æ–µ—Ç–æ—á–∏–µ–º, –∑–∞—Ç–µ–º —Å –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏ –∏–¥—ë—Ç –∫–æ–¥ —Å –æ—Ç—Å—Ç—É–ø–æ–º –≤ —á–µ—Ç—ã—Ä–µ –ø—Ä–æ–±–µ–ª–∞. –ü–æ—Å–ª–µ –±–ª–æ—á–Ω–æ–≥–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –º–∏–Ω–∏–º—É–º –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞ –≤–ª–æ–∂–µ–Ω–Ω–æ–≥–æ –∫–æ–¥–∞, –Ω–æ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–∫–æ–ª—å–∫–æ —É–≥–æ–¥–Ω–æ, –≤ —Ç–æ–º —á–∏—Å–ª–µ –¥—Ä—É–≥–∏–µ –±–ª–æ—á–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã.

### –í–∞–∂–Ω–æ: –µ—Å–ª–∏ `if` –º–æ–∂–µ—Ç –±—ã—Ç—å –±–µ–∑ `else`-—á–∞—Å—Ç–∏, —Ç–æ –≤–æ—Ç `else` –Ω–µ –º–æ–∂–µ—Ç —Å—É—â–µ—Å—Ç–≤–æ–≤–∞—Ç—å –±–µ–∑ —Å–≤–æ–µ–≥–æ `if`. –ï–¥–∞ –º–æ–∂–µ—Ç —Å—É—â–µ—Å—Ç–≤–æ–≤–∞—Ç—å –±–µ–∑ –∫–æ—Ç–∏–∫–∞, –∞ –≤–æ—Ç –∫–æ—Ç–∏–∫ –±–µ–∑ –µ–¥—ã ‚Äî –Ω–µ—Ç!

# –ë—É–ª–µ–≤–∞ –ª–æ–≥–∏–∫–∞
### –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã –æ—Ç–Ω–æ—à–µ–Ω–∏–π

## –í–æ–∫—Ä—É–≥ –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤ –æ—Ç–Ω–æ—à–µ–Ω–∏–π –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –Ω—É–∂–Ω–æ —Å—Ç–∞–≤–∏—Ç—å –ø–æ –ø—Ä–æ–±–µ–ª—É —Å –∫–∞–∂–¥–æ–π —Å—Ç–æ—Ä–æ–Ω—ã.

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ª—é–±–æ–π –∏–∑ –æ–ø–µ—Ä–∞—Ü–∏–π –æ—Ç–Ω–æ—à–µ–Ω–∏—è –º–æ–∂–µ—Ç –ø–æ–ª—É—á–∏—Ç—å—Å—è —Ç–æ–ª—å–∫–æ –¥–≤–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ ‚Äî –≤–µ—Ä–Ω–æ (–∏–ª–∏ –∏—Å—Ç–∏–Ω–Ω–æ) –∏–ª–∏ –Ω–µ–≤–µ—Ä–Ω–æ (–ª–æ–∂–Ω–æ). –í Python –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è (–∏ –µ—â—ë –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ñ–ª–∞–≥–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥—É—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Ç–æ–ª—å–∫–æ –¥–≤–∞ —Å–æ—Å—Ç–æ—è–Ω–∏—è ‚Äî –í–ö–õ/–í–´–ö–õ) –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö bool. –£ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö —ç—Ç–æ–≥–æ —Ç–∏–ø–∞ —Ç–æ–ª—å–∫–æ –¥–≤–∞ –∑–Ω–∞—á–µ–Ω–∏—è:

`True` ‚Äî ¬´–î–∞, —è —Å—ä–µ–ª –≤—Å—é –ø–∏—Ü—Ü—É¬ª.

`False` ‚Äî ¬´–ù–µ—Ç, —ç—Ç–æ –±—ã–ª –∫–æ—Ç!¬ª.

```python
print(2 == 2)  # True
print(3 < 1)  # False
print(5 <= 13)  # True
print("–ü—Ä–∏–≤–µ—Ç" == '–ø—Ä–∏–≤–µ—Ç')  # False ‚Äî —Ä–µ–≥–∏—Å—Ç—Ä –±—É–∫–≤ –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ!
```

–¥–≤–æ–π–Ω—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ –æ—Ç–Ω–æ—à–µ–Ω–∏—è:

```python
x_coord = int(input())
if 0 <= x_coord < 100:
    print('–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ –≤ –¥–æ–ø—É—Å—Ç–∏–º–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ')
```

## –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏–∫–∞–∫–∏—Ö –∫–∞–≤—ã—á–µ–∫

```python
is_pizza_gone = True  # –ò—Å—Ç–∏–Ω–Ω–æ –≤—Å—ë —Å—ä–µ–ª–∏
pizza_gone = 'True'  # –°—Ç—Ä–æ–∫–∞ True
print(is_pizza_gone == pizza_gone)  # –ö–æ–Ω–µ—á–Ω–æ –∂–µ, –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ —Ä–∞–≤–Ω—ã ‚Äî False
```

```python
is_cat_awake = 0
if is_cat_awake:  
    print("üò∏ –ú—É—Ä-–º—É—Ä!")  
else:  
    print("üí§ –ö–æ—Ç –≤ –æ—Ç–∫–ª—é—á–∫–µ.")
```

### –ó–∞–¥–∞—á–∞ ¬´–ö–æ—Ç–∏–∫ –≥–æ–ª–æ–¥–µ–Ω¬ª

–í Python –ø–æ—á—Ç–∏ –≤—Å—ë –º–æ–∂–Ω–æ –ø—Ä–µ–≤—Ä–∞—Ç–∏—Ç—å –≤ bool:

`False: 0`, –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ (–ø—Ä–∞–≤–∏–ª—å–Ω–µ–µ —Å–∫–∞–∑–∞—Ç—å ‚Äî –ª—é–±–∞—è –ø—É—Å—Ç–∞—è –∫–æ–ª–ª–µ–∫—Ü–∏—è);
    
`True`: –≤—Å—ë –æ—Å—Ç–∞–ª—å–Ω–æ–µ (–¥–∞–∂–µ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞! –∏ –¥–∞–∂–µ `0.000000001`!).
    
–≠—Ç–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —è–≤–Ω–æ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ `bool()`:

```python
print(bool('üê±'))  # True
print(bool(''))  # False
print(bool(0))  # False
print(bool(-3.14))  # True
```

–ù–æ –ø–æ—Ö–æ–∂–µ–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –Ω–µ—è–≤–Ω–æ –≤ –æ–ø–µ—Ä–∞—Ç–æ—Ä–µ `if` ‚Äî –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –Ω–µ –º–µ–Ω—è–µ—Ç—Å—è, –Ω–æ –≤–æ—Å–ø—Ä–∏–Ω–∏–º–∞–µ—Ç—Å—è –ø—Ä–æ–≥—Ä–∞–º–º–æ–π –∫–∞–∫ `True`/`False`:

```python
print("–ö–∞–∫ –∑–æ–≤—É—Ç –≤–∞—à–µ–≥–æ –∫–æ—Ç–∏–∫–∞?")
pet_name = input()
print("–ê —Å–∫–æ–ª—å–∫–æ –µ–º—É –ª–µ—Ç?")
pet_age = int(input())

if pet_name:  # –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤—ë–ª –ª—é–±–æ–µ –∏–º—è
    print(f"–ö–∞–∫–æ–µ –∫—Ä–∞—Å–∏–≤–æ–µ –∏–º—è ‚Äî {pet_name}")
else:  # –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø—Ä–æ—Å—Ç–æ –Ω–∞–∂–∞–ª Enter, –∏ –≤ pet_name ‚Äî –ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞
    print("–ë–µ–∑—ã–º—è–Ω–Ω—ã–π –∫–æ—Ç–∏–∫ ‚Äî —Ç–æ–∂–µ –∫–æ—Ç–∏–∫")

if pet_age:  # –õ—é–±–æ–µ –Ω–µ–Ω—É–ª–µ–≤–æ–µ —á–∏—Å–ª–æ
    print("–£–∂–µ –±–æ–ª—å—à–æ–π!")
else:  # –ö–æ—Ç–∏–∫—É 0 –ª–µ—Ç
    print('–ö–∞–∫–æ–π –º–∞–ª—ã—à!')
```

–î–æ–≤–æ–ª—å–Ω–æ —á–∞—Å—Ç–æ –±—ã–≤–∞–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç—å –≤–Ω—É—Ç—Ä–∏ –æ–¥–Ω–æ–≥–æ —É—Å–ª–æ–≤–∏—è –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –µ—â—ë –æ–¥–Ω–æ —É—Å–ª–æ–≤–∏–µ. –≠—Ç–æ –ª–µ–≥–∫–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å, —Ç–∞–∫ –∫–∞–∫ –≤–Ω—É—Ç—Ä–∏ –ª—é–±–æ–π —á–∞—Å—Ç–∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ `if-else` –º–æ–∂–µ—Ç –±—ã—Ç—å –¥—Ä—É–≥–æ–π –æ–ø–µ—Ä–∞—Ç–æ—Ä `if-else`, –∞ –≤–Ω—É—Ç—Ä–∏ –≤—Ç–æ—Ä–æ–≥–æ ‚Äî —Ç—Ä–µ—Ç–∏–π –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–ª–æ–∂–µ–Ω–Ω–æ—Å—Ç–∏ –º–∞—Ç—Ä—ë—à–∫–∏ —É—Å–ª–æ–≤–∏–π –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–æ —Ç–æ–ª—å–∫–æ –≤–∞—à–µ–π —Ñ–∞–Ω—Ç–∞–∑–∏–µ–π –∏ —Ç–µ—Ä–ø–µ–Ω–∏–µ–º.

```python
age = 25  
has_money = True  

if age >= 18:  
    if has_money:  
        print("–ú–æ–∂–Ω–æ –æ—Ç–¥–æ—Ö–Ω—É—Ç—å.")  
    else:  
        print("–ò—â–∏ —Ä–∞–±–æ—Ç—É! üíº")  
else:  
    print("–ò–¥–∏ —É—á–∏ —É—Ä–æ–∫–∏!")
```

### –ö–∞–∂–¥—ã–π –≤–ª–æ–∂–µ–Ω–Ω—ã–π –±–ª–æ–∫ –≤–Ω—É—Ç—Ä–∏ –≤–ª–æ–∂–µ–Ω–Ω–æ–≥–æ –±–ª–æ–∫–∞ –ø—Ä–∏–±–∞–≤–ª—è–µ—Ç –µ—â—ë 4 –ø—Ä–æ–±–µ–ª–∞ –∫ –æ—Ç—Å—Ç—É–ø—É. 

–ö–∞–∫ –Ω–µ—Å–ª–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, –µ—Å–ª–∏ —Å–∏–ª—å–Ω–æ —É–≤–ª–µ—á—å—Å—è –≤–ª–æ–∂–µ–Ω–Ω–æ—Å—Ç—å—é, –∫–æ–¥ –±—É–∫–≤–∞–ª—å–Ω–æ –±—É–¥–µ—Ç —É–±–µ–≥–∞—Ç—å –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –≤–∏–¥–∏–º–æ—Å—Ç–∏. –ß—Ç–æ–±—ã —ç—Ç–æ –Ω–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏–ª–æ –∏–ª–∏ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏–ª–æ —Ä–µ–∂–µ, –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –æ–ø–µ—Ä–∞—Ç–æ—Ä `elif`.

```python
age = int(input("–°–∫–æ–ª—å–∫–æ –ª–µ—Ç —Ç–µ–±–µ, –ø—É—Ç–Ω–∏–∫?"))

if age >= 3000:
    print("–û, –≤–µ–ª–∏–∫–∏–π –¥—Ä–µ–≤–Ω–∏–π! –†–∞—Å—Å–∫–∞–∂–∏ –ø—Ä–æ Python 0.1!")
elif age >= 100:
    print("–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ –º—É–∑–µ–π COBOL.")
else:
    print("–Æ–Ω–µ—Ü, –∏–¥–∏ —É—á–∏—Ç—å PEP8!")
```

# üß© –õ–æ–≥–∏—á–µ—Å–∫–æ–µ —Ç—Ä–∏–æ: not, and, or. –°–ª–µ–¥–∏—Ç–µ –∑–∞ —Ä—É–∫–∞–º–∏ (–∏ —É—Å–ª–æ–≤–∏—è–º–∏)

```python
if is_weekend or is_holiday:  
    print("–£—Ä–∞, –º–æ–∂–Ω–æ –≤–∞–ª—è—Ç—å—Å—è –≤ –≥–∞–º–∞–∫–µ!")
```

### –ß—Ç–æ–±—ã –ø–æ–≤—ã—Å–∏—Ç—å –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç –æ–ø–µ—Ä–∞—Ü–∏–π, –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å–∫–æ–±–∫–∏, —Ç–æ—á–Ω–æ —Ç–∞–∫ –∂–µ, –∫–∞–∫ –∏ –≤ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ.

–í–æ –º–Ω–æ–≥–∏—Ö —è–∑—ã–∫–∞—Ö –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è –µ—Å—Ç—å –æ–ø–µ—Ä–∞—Ç–æ—Ä –≤—ã–±–æ—Ä–∞ `switch`. –í Python –µ—Å—Ç—å –µ–≥–æ –∞–Ω–∞–ª–æ–≥ ‚Äî `match`/`case`. –ü—Ä–∏—á—ë–º —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª `match`/`case` –Ω–∞–º–Ω–æ–≥–æ –±–æ–≥–∞—á–µ, –∏ –º—ã –±—É–¥–µ–º –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å—Å—è –≤ –∫—É—Ä—Å–µ –∫ –Ω–µ–º—É –µ—â—ë –Ω–µ —Ä–∞–∑. –ù–æ –ø–æ–∫–∞ –º—ã —Ç–æ–ª—å–∫–æ –ø—Ä–∏–∫–æ—Å–Ω—ë–º—Å—è –∫ –µ–≥–æ –≤–µ–ª–∏—á–∏—é –∏ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –µ–≥–æ –¥–ª—è –∑–∞–º–µ–Ω—ã `if-elif-....-elif-else` –≤ —Å–ª—É—á–∞—è—Ö, –∫–æ–≥–¥–∞ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è –æ–¥–Ω–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è, –Ω–æ —É –Ω–µ—ë –º–æ–∂–µ—Ç –±—ã—Ç—å –º–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–π:

```python
day = input()  
match day:  # –°–º–æ—Ç—Ä–∏–º, –∫–∞–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É day
    case "–ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫":  # if day == "–ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫"
        print("–ù–∞—á–∞–ª–æ —Ä–∞–±–æ—á–µ–π –Ω–µ–¥–µ–ª–∏...")  
    case "—Å—É–±–±–æ—Ç–∞" | "–≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ":  # if day == "—Å—É–±–±–æ—Ç–∞" or day =="–≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ"
        print("–£–∏–∏–∏–∏! –í—ã—Ö–æ–¥–Ω—ã–µ! üéâ")  
    case _:  # else
        print("–õ—é–±–æ–π –¥—Ä—É–≥–æ–π –¥–µ–Ω—å.") 
```

* `|` ‚Äî –ª–æ–≥–∏—á–µ—Å–∫–æ–µ –ò–õ–ò;
* `_` ‚Äî –¥–µ—Ñ–æ–ª—Ç–Ω—ã–π –∫–µ–π—Å (–∫–∞–∫ else).

```python
pizza_pieces = int(input())  
match pizza_pieces:  
    case 0:  
        print("–ú–Ω–µ –Ω–µ —Ö–≤–∞—Ç–∏–ª–æ(")
    case 1:
        print("–°–ø–∞—Å–∏–±–æ, —á—Ç–æ –æ—Å—Ç–∞–≤–∏–ª–∏, –Ω–æ –º–∞–ª–æ.")  
    case 2 | 3 | 4:  
        print("–≠—Ç–æ –ø–æ –º–Ω–µ!")
        print("–£—Ä–∞!")
    case _:  
        print("–°—ä–µ–º –≤–æ —á—Ç–æ –±—ã —Ç–æ –Ω–∏ —Å—Ç–∞–ª–æ!")
```
